### –¢–µ–æ—Ä–µ–º–∞:

### –°–ª–µ–¥—Å—Ç–≤–∏–µ:

### –ê–ª–≥–æ—Ä–∏—Ç–º –ë—Ä–∞—É—ç—Ä–∞:

–î–ª—è $n \in ‚Ñï$ –ø—Ä–∏ –∑–∞–¥–∞–Ω–Ω–æ–º $k \in ‚Ñï$ –º–æ–∂–Ω–æ –ø–æ—Å—Ç—Ä–æ–∏—Ç—å —Ü–µ–ø–æ—á–∫—É –ë—Ä–∞—É—ç—Ä–∞ —Å –ø–æ–º–æ—â—å—é —Ä–µ–∫—É—Ä—Ä–µ–Ω—Ç–Ω–æ–π —Ñ–æ—Ä–º—É–ª—ã:

$$ B_k (n) =
\begin{cases}

B_k (q), 2q, 4q, ..., 2^k q, n, \quad n \ge 2^k
\end{cases} \\

$$

–î–∞–Ω–Ω–∞—è —Ü–µ–ø–æ—á–∫–∞ –∏–º–µ–µ—Ç –¥–ª–∏–Ω—É
$$l_B (n) = j(k + 1) + 2^k - 2,$$
–ø—Ä–∏ —É—Å–ª–æ–≤–∏–∏ —á—Ç–æ $jk \le \lambda (n) \le (j+1)k$

–î–ª–∏–Ω–∞ –±—É–¥–µ—Ç –º–∏–Ω–∏–º–∞–ª–∏–∑–∏—Ä–æ–≤–∞–Ω–∞ –¥–ª—è –±–æ–ª—å—à–∏—Ö $n$, –µ—Å–ª–∏ –ø–æ–ª–æ–∂–∏—Ç—å $k = \lambda \lambda (n) - 2 \lambda \lambda \lambda (n)$

**–•–æ–¥ –∞–ª–≥–æ—Ä–∏—Ç–º–∞:**

* –ó–∞–¥–∞—ë—Ç—Å—è –Ω–µ–∫–∏–π –ø–∞—Ä–∞–º–µ—Ç—Ä $k$ –¥–ª—è $n$.
–í—ã–ø–æ–ª–Ω—è–µ—Ç—Å—è –≤—ã—á–∏—Å–ª–µ–Ω–∏–µ –≤—Å–ø–æ–º–æ–≥–∞—Ç–µ–ª—å–Ω—ã—Ö —á–∏—Å–µ–ª:
$$
d = 2^k, \hspace{0.2cm} q_1 = [ {n \over d} ], \hspace{0.2cm} r_1 = n \hspace{0.2cm} mod \hspace{0.2cm} d => n = q_1 d + r_1 \quad (0 \le r_1 < d) \\
q_2 = [ {q_1 \over d} ], \hspace{0.2cm} r_2 = q_1 \hspace{0.2cm} mod \hspace{0.2cm} d => q_1 = q_2 d + r_2 \quad (0 \le r_2 < d)
$$
* –î–∞–Ω–Ω–∞—è –ø—Ä–æ—Ü–µ–¥—É—Ä–∞ –ø—Ä–æ–¥–æ–ª–∂–∞–µ—Ç—Å—è, –ø–æ–∫–∞ –Ω–µ –ø–æ—è–≤–∏—Ç—Å—è $q_s < d,$ —Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ $q_{s-1} = q_s d + r_s$
* –¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º, n –∏–º–µ–µ—Ç –≤–∏–¥
$$ n = 2^k q_1 + r_1 = 2^k (2^k q_2 + r_2) + r_1 = ...\\
... = 2^k (2^k (... (2^k q_s + r_s ) ... ) + r_2 ) + r_1 . $$

$$B_n (n): 1, 2, 3, ..., 2^k - 1, \\
2q_s, 4q_s, 8q_s, ..., 2^k q_s, 2^k q_s + r_s, \\
2q_{s-1}, 4q_{s-1}, 8q_{s-1}, ..., 2^k q_{s-1}, 2^k q_{s-1} + r_{s-1}, \\

### –ê–ª–≥–æ—Ä–∏—Ç–º –Ø–æ:

* –û–±–ª–∞–¥–∞–µ—Ç —Ç–∞–∫–æ–π –∂–µ –≤—ã—á–∏—Å–ª–∏—Ç–µ–ª—å–Ω–æ–π –º–æ—â–Ω–æ—Å—Ç—å—é, —á—Ç–æ –∏ –∞–ª–≥–æ—Ä–∏—Ç–º –ë—Ä–∞—É—ç—Ä–∞

### 2. –†–µ–∞–ª–∏–∑–æ–≤–∞—Ç—å –∞–ª–≥–æ—Ä–∏—Ç–º –ë—Ä–∞—É—ç—Ä–∞ –¥–ª—è –≤—ã—á–∏—Å–ª–µ–Ω–∏—è –ø—Ä–∏–±–ª–∏–∂—ë–Ω–Ω—ã—Ö –∞–¥–¥–∏—Ç–∏–≤–Ω—ã—Ö —Ü–µ–ø–æ—á–µ–∫ –¥–ª—è —Ä–∞–∑–ª–∏—á–Ω—ã—Ö —á–∏—Å–µ–ª –ø—Ä–∏ –≤–∞—Ä—å–∏—Ä–æ–≤–∞–Ω–∏–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ ùëò, —Å–æ–ø–æ—Å—Ç–∞–≤–∏—Ç—å –¥–ª–∏–Ω—ã –ø–æ–ª—É—á–µ–Ω–Ω—ã—Ö –∞–¥–¥–∏—Ç–∏–≤–Ω—ã—Ö —Ü–µ–ø–æ—á–µ–∫ —Å –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–π –∞–¥–¥–∏—Ç–∏–≤–Ω–æ–π —Ü–µ–ø–æ—á–∫–æ–π –¥–ª—è –∑–∞–¥–∞–Ω–Ω–æ–≥–æ —á–∏—Å–ª–∞. –°–¥–µ–ª–∞—Ç—å –≤—ã–≤–æ–¥—ã. 

```python
def _lambda(n):
    n = Integer(n)
    return n.nbits() - 1

def min_chain_len(n):
    return int(_lambda(n) + _lambda(n) / _lambda(_lambda(n)) + (_lambda(n) * _lambda(_lambda(_lambda(n)))) / (_lambda(_lambda(n)) ** 2))


def to_str(chain):
    return " ".join(map(str, chain[:min(len(chain), 17)])) + (" ..." if len(chain) > 17 else "")

def brauer(n, k):
    n = Integer(n)
    
    d = 2 ** k
    
    rarr = [] # array of "remainders" (d-ary number digits)
    q = n.quo_rem(d)[0]
    rarr.append(n.quo_rem(d)[1])
    sqmax = rarr[0] # maximum number in the digits of n

    i = 0
    while q >= d:
        rarr.append(q.quo_rem(d)[1])
        q = q.quo_rem(d)[0]
        if rarr[i] > sqmax:
            sqmax = rarr[i]
        i += 1
    rarr.append(q)
    sqmax = max(q, sqmax)

    chain = Sequence([0]*(sqmax + (len(rarr) - 1) * (k + 1)), ZZ) # addition chain
    for i in range(1, sqmax + 1):
        chain[i-1] = i # starting sequence 1, 2, 3, ..., b, where b is maximum number in the digits of n

    chain[sqmax - 1] = q
    for i in range(0, len(rarr) - 1): # i - digit number
        for j in range(0, k): # j - number to double
            chain[sqmax + i * (k + 1) + j] = chain[sqmax + i * (k + 1) + j - 1] * 2
        chain[sqmax + i * (k + 1) + k] = chain[sqmax + i * (k + 1) + k - 1] + rarr[len(rarr) - 2 - i]
    chain[sqmax - 1] = sqmax
    
    chain = sorted(set(chain))
    
    if chain[0] == 0:
        chain = chain[1:]

    return chain

set_random_seed(352073568)
n_set = Sequence([0]*10, ZZ)

for i in range(0, 5):
    n_set[i] = randint(17, 127)
for i in range(0, 5):
    n_set[i+5] = randint(127, 65535)
    
for k in [1, 2, 3, 6, 9, 16]:
    print("k = ", k, ":")
    for n in n_set:
        chain = brauer(n, k)
        print("   n = {:5};  chain_len(n) = {:5};  l(n) = {:5}  | {}".format(n, len(chain), min_chain_len(n), to_str(chain)))
```

–ò–∑ –ø—Ä–æ–¥–µ–ª–∞–Ω–Ω—ã—Ö –æ–ø—ã—Ç–æ–≤ –≤–∏–¥–Ω–æ, —á—Ç–æ –æ–ø—Ç–∏–º–∞–ª—å–Ω–æ–µ $k(n)$ —Å–æ–≤—Å–µ–º –Ω–µ–≤–µ–ª–∏–∫–æ –ø–æ —Å—Ä–∞–≤–Ω–µ–Ω–∏—é —Å $n$ –¥–ª—è –±–æ–ª—å—à–∏—Ö $n$, –±–æ–ª–µ–µ —Ç–æ—á–Ω–æ –æ–Ω–æ –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç—Å—è –∫–∞–∫
$$ k = \lambda \lambda (n) - 2 \lambda \lambda \lambda (n) $$

–ó–Ω–∞—á–µ–Ω–∏—è $n$, –≤–∑—è—Ç—ã–µ –¥–ª—è –æ–ø—ã—Ç–æ–≤, –±—ã–ª–∏ –Ω–µ–¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ –±–æ–ª—å—à–∏–º–∏, —á—Ç–æ–±—ã –æ–ø—Ç–∏–º–∞–ª—å–Ω–æ–µ $k$ —É–≤–µ–ª–∏—á–∏–≤–∞–ª–æ—Å—å –¥–∞–ª–µ–µ 2. –¢.–µ. –≤ –Ω–∞—à–µ–º —Å–ª—É—á–∞–µ –æ–Ω–æ –±—ã–ª–æ –º–µ–Ω—å—à–µ –∏–ª–∏ —Ä–∞–≤–Ω–æ $n$. –≠—Ç–æ –≤–∏–¥–Ω–æ, —Ç.–∫. –ø—Ä–∏ $k$ –Ω–µ –ø—Ä–µ–≤—ã—à–∞—é—â–µ–º 2 –¥–ª–∏–Ω—ã —Ü–µ–ø–æ—á–µ–∫ –Ω–µ —É–≤–µ–ª–∏—á–∏–≤–∞–ª–∏—Å—å. 

–†–∞—Å—Å–º–æ—Ç—Ä–µ–≤ –º–∞–ª—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è $k$, —Ä–∞—Å—Å–º–æ—Ç—Ä–∏–º –±–æ–ª—å—à–∏–µ. –í–∏–¥–Ω–æ, —á—Ç–æ —Å–ª–∏—à–∫–æ–º –±–æ–ª—å—à–æ–µ $k$ –≤–µ–¥—ë—Ç –∫ –¥–µ–≥—Ä–∞–¥–∞—á–∏–∏ —Ü–µ–ø–æ—á–µ–∫ –¥–æ —Ä—è–¥–æ–≤ –Ω–∞—Ç—É—Ä–∞–ª—å–Ω—ã—Ö —á–∏—Å–µ–ª. –ü–æ–¥–æ–±–Ω–æ–µ –ø–æ–≤–µ–¥–µ–Ω–∏–µ –Ω–∞–±–ª—é–¥–∞–µ—Ç—Å—è —É–∂–µ –ø—Ä–∏ $k = 6$. –ü–æ–ª–Ω–æ–µ –∂–µ —Å–æ–≤–ø–∞–¥–µ–Ω–∏–µ —Ü–µ–ø–æ—á–∫–∏ –º–µ—Ç–æ–¥–∞ –ë—Ä–∞—É—ç—Ä–∞ —Å —Ä—è–¥–æ–º –Ω–∞—Ç—É—Ä–∞–ª—å–Ω—ã—Ö —á–∏—Å–µ–ª –ø—Ä–æ–∏—Å—Ö–æ–¥–∏—Ç –ø—Ä–∏ —Ç–∞–∫–æ–º $k$, —á—Ç–æ $2^k \ge n$. –≠—Ç–æ –º–æ–∂–Ω–æ —É–≤–∏–¥–µ—Ç—å –ø—Ä–∏ $k = 16$, –≤–µ–¥—å $2^{16} = 65536$, —á—Ç–æ –±–æ–ª—å—à–µ –ª—é–±–æ–≥–æ –≤—ã–±—Ä–∞–Ω–Ω–æ–≥–æ $n$.

–¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º, —á—Ç–æ–±—ã –∞–ª–≥–æ—Ä–∏—Ç–º –ë—Ä–∞—É—ç—Ä–∞ —Å—Ç—Ä–æ–∏–ª –∞–¥–¥–∏—Ç–∏–≤–Ω—ã–µ —Ü–µ–ø–æ—á–∫–∏ –∞–¥–µ–∫–≤–∞—Ç–Ω–æ–π –¥–ª–∏–Ω—ã, $k$ –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å —Å–∏–ª—å–Ω–æ –º–µ–Ω—å—à–µ $n$ –∏ –¥–ª—è –¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ –±–æ–ª—å—à–∏—Ö $n$ –≤—ã—á–∏—Å–ª—è—Ç—å—Å—è –ø–æ —Ñ–æ—Ä–º—É–ª–µ $ k = \lambda \lambda (n) - 2 \lambda \lambda \lambda (n) $, –∞ –¥–ª—è –Ω–µ–±–æ–ª—å—à–∏—Ö $n$ –±—ã—Ç—å –ø–æ–ª–æ–∂–µ–Ω–æ —Ä–∞–≤–Ω—ã–º $1$ –∏–ª–∏ $2$.

### 3. –†–µ–∞–ª–∏–∑–æ–≤–∞—Ç—å –∞–ª–≥–æ—Ä–∏—Ç–º –¥—Ä–æ–±–ª–µ–Ω–∏—è –≤–µ–∫—Ç–æ—Ä–∞ –∏–Ω–¥–µ–∫—Å–æ–≤ –¥–ª—è –Ω–∞—Ö–æ–∂–¥–µ–Ω–∏—è –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–π –∑–≤—ë–∑–¥–Ω–æ–π —Ü–µ–ø–æ—á–∫–∏ –¥–ª—è –∑–∞–¥–∞–Ω–Ω–æ–≥–æ —á–∏—Å–ª–∞. –ü—Ä–æ—Ç–µ—Å—Ç–∏—Ä–æ–≤–∞—Ç—å –∞–ª–≥–æ—Ä–∏—Ç–º –º–∏–Ω–∏–º—É–º –¥–ª—è 5 –∑–Ω–∞—á–µ–Ω–∏–π n > 1000. –£–∫–∞–∑–∞—Ç—å, —Å–∫–æ–ª—å–∫–æ –≤—Ä–µ–º–µ–Ω–∏ –ø–æ—Ç—Ä–µ–±–æ–≤–∞–ª–æ—Å—å –Ω–∞ –ø–æ–∏—Å–∫ —Ü–µ–ø–æ—á–∫–∏ –∏ –∫–∞–∫–∞—è —Ü–µ–ø–æ—á–∫–∞ –ø–æ–ª—É—á–∏–ª–∞—Å—å. –°—Ä–∞–≤–Ω–∏—Ç—å —Å –ø—Ä–µ–¥—ã–¥—É—â–∏–º–∏ –º–µ—Ç–æ–¥–∞–º–∏, —Å–¥–µ–ª–∞—Ç—å –≤—ã–≤–æ–¥—ã.

```python
from sage.functions.log import logb
import time

def _lambda(n):
    n = Integer(n)
    return n.nbits() - 1

def min_chain_length(n):
    return _lambda(n) + 1

def decrement_vector(vector, start_index, end_index, delta=1):
    index = end_index
    while index >= start_index and vector[index] == 1:
        index -= 1
    if index < start_index:
        return False
    vector[index] -= 1
    for i in range(index + 1, end_index + 1):
        vector[i] = i + delta
    return True


def calculate_chain_by_vector(vector):
    chain = [1]
    for i in range(0, len(vector)):
        chain.append(chain[-1] + chain[vector[i] - 1])
    return chain


def chain_from_part_of_vector(part, chain_part):
    chain = chain_part[:]
    for i in range(0, len(part)):
        chain.append(chain[-1] + chain[part[i] - 1])
    return chain


def indexes_vector_division(number):
    m = ceil(logb(number, 2))
    while True:
        if 2**m == number:
            return list(range(1, m + 1))
        if 2**m < number:
            m += 1
            continue
        result = crush_part_of_vector(number, list(range(1, m+1)))
        if result == False:
            m += 1
        else:
            return result
            
            
def crush_part_of_vector(number, part, first_max_value = 1, chain = [1]):
    m = len(part)
    if len(part) == 1:
        while part[0] > 1:  
            a = chain_from_part_of_vector(part, chain)
            a = a[-1]
            if a < number:
                return False
            if number == a:
                return part
            part[0] -= 1
        return False
    q = m // 2
    can_dec_fixed_part = True
    while can_dec_fixed_part:
        all_chain = chain_from_part_of_vector(part, chain)
        a = all_chain[q + first_max_value - 1]
        a_min = a + m - q
        a_max = a * 2 ** (m - q)
        if a_max < number:
            return False
        if a_min == number:
            return part[:q] + [1 for _ in range(q + 1, m + 1)]
        if a_max == number:
            return part
        if a_min < number < a_max:
            crush = crush_part_of_vector(number, part[q:], first_max_value + q, all_chain[:q + first_max_value])
            
            if crush:
                return part[:q] + crush
        can_dec_fixed_part = decrement_vector(part, 0, q - 1, first_max_value)
    return False
        

set_random_seed(352073568)

numbers = []

for i in range(0, 5):
    numbers.append(randint(1001, 3000))

for n in numbers:
    print("n = ", n)
    
    start_point = time.time() 
    
    vector = indexes_vector_division(n)
    print("\t Indexes: ", vector)
    
    chain = calculate_chain_by_vector(vector)
    print("\t Chain: ", chain)
    
    print("\t Chain length: ", len(chain))
    print("\t Min chain length: ", min_chain_len(n))
    
    end_point = time.time()
    
    print("\t Time elapsed: ", end_point - start_point)
    
    print("")
```

### 5. –ü—Ä–æ–≤–µ—Ä–∏—Ç—å –≥–∏–ø–æ—Ç–µ–∑—É –®–æ–ª—å—Ü–∞‚Äì–ë—Ä–∞—É—ç—Ä–∞ –¥–ª—è –≤—Å–µ—Ö –Ω–∞—Ç—É—Ä–∞–ª—å–Ω—ã—Ö 1 $\leq$ ùëõ $\leq$ 12 –Ω–∞ –∞–ª–≥–æ—Ä–∏—Ç–º–µ –¥—Ä–æ–±–ª–µ–Ω–∏—è –≤–µ–∫—Ç–æ—Ä–∞ –∏–Ω–¥–µ–∫—Å–æ–≤. –†–µ–∑—É–ª—å—Ç–∞—Ç—ã –æ—Ñ–æ—Ä–º–∏—Ç—å –≤ –≤–∏–¥–µ —Ç–∞–±–ª–∏—Ü—ã. –°–¥–µ–ª–∞—Ç—å –≤—ã–≤–æ–¥—ã.

```python
for n in range(1, 13):
    start = time.time()
    
    number = 2**n - 1
    l1 = indexes_vector_division(number)
    l2 = indexes_vector_division(n)
    
    end = time.time()
    
    print("n =", n)
    print("\tNumber =", number)
    print("\tChain for " + str(number) + ":", calculate_chain_by_vector(l1))
    print("\tChain for " + str(n) + ":", calculate_chain_by_vector(l2))
    print("\tTime elapsed:", end - start, "sec")
    if len(l1) <= len(l2) + n - 1:
        print("\tTheorem: PASS\t", len(l1), '<=', len(l2) + n - 1)
        print()
    else:
        print("\tTheorem: FAILED")
        print(len(l1), ' > ', len(l2) + n - 1)
        print(l1)
        break
```
