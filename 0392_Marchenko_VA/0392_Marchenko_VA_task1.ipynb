## Основные теоретические положения


**Бинарный метод**

Запишем n в двоичной системе счисления и заменим в этой записи каждую цифру “1” парой букв SX, а каждую цифру “0” – буквой S, после чего вычеркнем крайнюю левую пару букв SX. Результат, читаемый слева на право, превращается в правило вычисления хn, если букву “S” интерпретировать как операцию возведения в квадрат (S – square – квадрат), а букву “X ” – как операцию умножения на х.

**Метод множителей**
Если n=p·q, где р – наименьший простой множитель числа n и q >1, то для вычисления хn мы можем сначала вычислить хр, а затем возвести это число в степень q. В случае, когда n простое, мы можем вычислить сначала число хn-1, а затем умножить его на х.

Если n=1, число хn имеется без всяких вычислений. Многократное применение этих правил дает нам процедуру вычисления хn для любого данного n.

**Аддитивная цепочка** — последовательность натуральных чисел, начинающаяся с единицы, в которой каждый последующий элемент является суммой каких-то двух предшествующих элементов (в том числе, возможно использование одного и того же предшествующего элемента — удвоение).



**Алгоритм Брауэра**

1. Задаётся некий $k$ для числа $n$, выч:
$$ 
d = 2^k, \hspace{0.2cm} q_1 = [ {n \over d} ], \hspace{0.2cm} r_1 = n \hspace{0.2cm} mod \hspace{0.2cm} d \rightarrow n = q_1 d + r_1 \quad (0 \le r_1 < d) \\
q_2 = [ {q_1 \over d} ], \hspace{0.2cm} r_2 = q_1 \hspace{0.2cm} mod \hspace{0.2cm} d \rightarrow q_1 = q_2 d + r_2 \quad (0 \le r_2 < d)
$$
1. Продолжаем, пока не появится $q_s < d.$ Следовательно, $q_{s-1} = q_s d + r_s$
1. Таким образом:
$$ n = 2^k q_1 + r_1 = 2^k (2^k q_2 + r_2) + r_1 = ... = 2^k (2^k (... (2^k q_s + r_s ) ... ) + r_2 ) + r_1 $$

$
B_k (n) = 
\begin{cases} 1, 2, 3, ..., 2^k - 1, если n < 2^k
\\B_k(q), 2q, 4q, 8q, ..., 2^k q, если n \geq 2^k
\end{cases}
$


**Звездный шаг** — шаг в аддитивной цепочке, при котором новый элемент получается сложением предыдущего с каким-либо.


**Звездная цепочка** — аддитивная цепочка, в которой все шаги звездные.

**l*(n)** - длина минимальной звёздной цепочки.


Для любой звёздной цепочки можно задать вектор индексов. Вектором индексов аддитивной цепочки называется последовательность номеров чисел в аддитивной цепочке, которые мы прибваляем к последнему числу


$
{[r_i]^k}_{i=0}
$

где 
$
r_i = [x:1 <= x <= i]
$

**Алгоритм дробления вектора индексов**
Алгоритм вычисляет min l*(n)

1)Строим базовый вектор


2)Разбиваем на 2 части: фиксированную и меняющуюся 


3)Вычисляем максимальное a_max и минимальное число a_min, которое мы можем получить при текущей фикс. части


  3.1) Если n = a_max/a_min то задача решена
  
  
  3.2) Если n < a_min или n > a_max, то меняем фикc. часть
  
  
  3.3) Если a_min < n < a_max, перебераем меняющуюся часть и находим ответ
  
  
  
4)Если пункт 3 не сработал увеличиваем длину


**Гипотеза Шольца-Брауэра**



$$
l*(2^n - 1) >= l*(n) + n - 1
$$

```python
def brouwer(x, k):
    d = 2 ** k
    chain = list(range(1, d))
    rems = []
    
    y = copy(x)
        
    while y > 0:
        u = y//d 
        if u:
            rems.append(y%d)
            y = u
        else:
 
            break
        
    rems = list(reversed(rems))

    chain.append(y)

    
    for i in rems:
        q=chain[-1]
        
        for j in range(k):
            chain.append(2**(j+1)*q)
            
      
        chain.append(chain[-1] + i)
        
    return list(sorted(set(chain)))

    
```

## III
**Алгоритм дробления вектора индексов**


```python
import math
import time
from decimal import *

def lmbda(n):
    return len(n.digits(2))-1

def nu(n):
    return bin(n).count('1')
    
def star_chain_by_vec(vec):
    ch = [1]
    
    for i in vec:
        ch.append(ch[-1] + ch[i - 1])
    
    return ch

def reduceVector(vector, q):
    i = 1
    l = len(vector)
    while vector[-i] == 1:
        if i != l:
            vector[-i] = l - i + 1 + q
            i += 1
        else:
            break
    if i != l or vector[-i] != 1:
        vector[-i] -= 1


def split_vec(n):
    prev_chain = []
    if n == 3:
        return [1, 1]
    
    niz = lmbda(n)
    verh = lmbda(n) + nu(n) - 1
               
    for m in range(niz, verh + 1):
        flag = True
        q = m // 2
        vec = list(range(1, m + 1))
        s_chain = star_chain_by_vec(vec)
        fixated = vec[:q]
        changeable = vec[q:]
        a = s_chain[q]
        a_min = a + m - q
        a_max = a *(2**(m-q))
        while flag:
            
            if n < a_min or n > a_max:
                
                if q > 1:
                    q -= 1  

                    fixated = vec[:q]
                    changeable = vec[q:]
                    s_chain = star_chain_by_vec(fixated + changeable)
                    a = s_chain[q]
                    a_min = a + m - q
                    a_max = a *(2**(m-q))
                else:
                    break
                
            else:               
                while True:
                    
                    chain = star_chain_by_vec(fixated + changeable)

                    if chain == prev_chain:
                        flag = False
                        break
                    prev_chain = chain

                    if chain[-1] == n:
                        return fixated + changeable
                    
                    if chain[-1] < n:
                        changeable[-1] = 1
                        if changeable.count(1) != len(changeable):
                            reduceVector(changeable, q)
                            
                    else:
                        reduceVector(changeable, q)
                    
                    chain = star_chain_by_vec(fixated + changeable)
        reduceVector(fixated, 0)
        changeable = vec[q:]
                    

            
def manager(n):
    t1 = time.time()
    v = split_vec(n)
    time_delta =  Decimal(time.time() - t1)
    c = star_chain_by_vec(v)
    
    print(f"Для n = {n} \nвектор будет:")
    print(v)
    print("А цепочка соответсвенно: ")
    print(c)
    print(f"С длиной l = {len(c) - 1}")
    print("На вычисление алгоритму понадобилось")
    print(time_delta)
    print("секунд")
    print("------------------------------")
    print()
    

```

```python

manager(510)
manager(514)


manager(1000)
manager(1011)
manager(1023)
manager(1024)
manager(1048)


manager(1600)
```
