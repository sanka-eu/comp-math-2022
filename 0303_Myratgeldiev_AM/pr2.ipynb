# Основные теоретические положения

## Группа 
Множество $G$ с бинарной операцией $\cdot$ называется группой, если: операция $\cdot$ ассоциативна; в $G$ существует нейтральный элемент $e$; для любого $a \in G$ существует обратный элемент, т.е. такой элемент $a' \in G$, что $ a \cdot a' = a' \cdot a = e$

## Циклическая группа
Группа, порожденная одним элементом, называется циклической группой и обозначается $<a>$.

Пусть $G$ - группа с нейтральным элементом $e$, а $X$ - множество. Будем говорить, что $G$ действует на $X$, если задана операция: $G\times X \rightarrow X$, обладающая для любого $x$ $\in$ $X$ и $g$, $h$ $\in$ $G$ следующими свойствами

##### 1. $e \cdot x = x$
##### 2. $g \cdot h \cdot x = g(h(x))$

## Орбита
Орбитой элемента $x \in X$ под дествием $G$ называется множество $Gx = \{gx: g \in G\}$.
Число $\#(Gx)$ называют длиной орбиты.

## Неподвижные точки
Неподвижными точками элемента $g \in G$ называются точки $x \in X$, для которых $gx = x$. Множество неподвижных точек элемента $g$ обозначается $X^g$.

## Лемма Бёрнсайда
Количество орбит действия группы $G$ на множестве $X$ равно: 
$$N = \dfrac{1}{\# G}\displaystyle\sum\limits_{g \in G} \#X^g$$

Пусть $I$ - произвольное множество, а C - множество цветов. Расскраской множества $I$ называется функция из $I$ в $C$.

Множество всех раскрасок обозначается $C^I$.

Количество всех раскрасок равно $\#(C^I)$

Количество раскрасок из $C^I$, которое сохраняет данный элемент $g \in G$ равно $(\#C)^{c(\phi(g))}$.

Здесь $c(\sigma)$ обозначает количество независимых циклов в записи перестановки $\sigma$

### 1. Аналитическое решение задачи
Введем обозначения:
Пусть $\sigma$ - поворот куба на 90 градусов по часовой стрелке, $\tau$ - переворот куба на лицевую сторону, а $\psi$ - переворот куба на правую сторону.

Тогда запишем эти операции в виде перестановок:

$\sigma = 
\begin{pmatrix}
  1~2~3~4~5~6~7~8\\
  4~1~2~3~8~5~6~7\\
\end{pmatrix} = (1\ 4\ 3\ 2)(5\ 8\ 7\ 6)$

$\tau = 
\begin{pmatrix}
  1~2~3~4~5~6~7~8\\
  2~6~7~3~1~5~8~4\\
\end{pmatrix} = (1\ 2\ 6\ 5)(3\ 7\ 8\ 4)$

$\psi = 
\begin{pmatrix}
  1~2~3~4~5~6~7~8\\
  5~6~2~1~8~7~3~4\\
\end{pmatrix} = (1\ 5\ 8\ 4)(2\ 6\ 7\ 3)$

```python
sigma = "(1, 4, 3, 2)(5, 8, 7, 6)" # поворот на 90 градусов по часовой стрелке
tau = "(1, 2, 6, 5)(3, 7, 8, 4)" # переворот на лицевую сторону
psi = "(1, 5, 8, 4)(2, 6, 7, 3)" # переворот на правую сторону
G = PermutationGroup([sigma, tau, psi]) # всевозможные комбинации sigma, tau, psi
print('Пусть G - множество всех комбинаций поворотов куба. Тогда, |G| = 24. Все элементы G:')
for i in range(G.order()):
    print('$$', i+1, '\\rightarrow', G[i].cycle_string(singletons=True), '$$')
```

### 2. Графическое отображение раскраски по заданному номеру
Реализуем функцию, которая будет отображать раскраску по заданному номеру и по заданным цветам

```python
def colorCube(num, colors=['red', 'green', 'blue', 'orange', 'cyan']):
    # координаты вершин куба
    v_i = [(0, 1), (0.4, 1.4), (1.4, 1.4), (1, 1), (0, 0), (0.4, 0.4), (1.4, 0.4), (1, 0)]
    # отрезки, соединяющие соответствующе вершины
    lines = [(v_i[0], v_i[1]), (v_i[1], v_i[2]), (v_i[2], v_i[3]), (v_i[0], v_i[3]), (v_i[0], v_i[4]), (v_i[4], v_i[5]), (v_i[4], v_i[7]), (v_i[5], v_i[1]), (v_i[5], v_i[6]), (v_i[6], v_i[2]), (v_i[6], v_i[7]), (v_i[7], v_i[3])]
    G = Graphics()
    for i in lines:
        G += line2d(i, color='black', thickness=0.3)
    for i in range(8):
        G += point2d(v_i[i], color=colors[num % len(colors)], size=150, axes=False)
        G += text(str(i+1), (v_i[i][0]+0.05, v_i[i][1]), color=colors[num % len(colors)], fontsize=12)
        num //=  len(colors)
    return G
```

Сколькими способами можно раскрасить вершины куба используя краски k цветов (с точностью до поворотов куба).

В нашем задании, $k = 5$. Тогда в формулу, по которой мы вычислили количество раскрасок, вместо 5 подставим k. Тогда, получим:
\\[N = \dfrac{1}{24} \cdot (k^8 + 17 \cdot k^4 + 6 \cdot k^2) = \dfrac{k^8 + 17 \cdot k^4 + 6 \cdot k^2}{24}\\]

Проверим, формулу для нескольких k:

При $k = 1$:
\\[N = \dfrac{1}{24} \cdot (1^8 + 17 \cdot 1^4 + 6 \cdot 1^2) = \dfrac{1 + 17 + 6}{24} = 1\\]
Действительно, при 1 цвете, вершины куба можно раскрасить единственным способом.

При $k = 3$:
\\[N = \dfrac{1}{24} \cdot (3^8 + 17 \cdot 3^4 + 6 \cdot 3^2) = 333\\]

При $k = 8$:
\\[N = \dfrac{1}{24} \cdot (8^8 + 17 \cdot 8^4 + 6 \cdot 8^2) = 701968\\]
