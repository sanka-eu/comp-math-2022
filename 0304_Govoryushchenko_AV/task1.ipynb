$$x^{67} = x*x^{66} = x*(x^{33})^2 = x*((x^{11})^3)^2 = x*(x^{11}*(x^{11})^2)^2 , $$ –≥–¥–µ $$x^{11} = x*x^{10} = x*(x^5)^2 = x*(x*(x^2)^2)^2$$

<p>–ò—Å—Ö–æ–¥—è –∏–∑ —Ç–∞–±–ª–∏—Ü—ã –º–æ–∂–Ω–æ —Å–¥–µ–ª–∞—Ç—å –≤—ã–≤–æ–¥, —á—Ç–æ –±–∏–Ω–∞—Ä–Ω—ã–π –º–µ—Ç–æ–¥ —Ä–∞–±–æ—Ç–∞–µ—Ç —Ö—É–∂–µ –≤ —Ç–µ—Ö —Å–ª—É—á–∞—è—Ö, –∫–æ–≥–¥–∞ –≤ –¥–≤–æ–∏—á–Ω–æ–π –∑–∞–ø–∏—Å–∏ —á–∏—Å–ª–∞ —Å–æ–¥–µ—Ä–∂–∏—Ç—å—Å—è –±–æ–ª—å—à–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –µ–¥–∏–Ω–∏—Ü. –ó–∞–≤–∏—Å–∏–º–æ—Å—Ç—å —á–∏—Å–ª–∞ –æ–ø–µ—Ä–∞—Ü–∏–π –æ—Ç –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –µ–¥–∏–Ω–∏—Ü –≤ –¥–≤–æ–∏—á–Ω–æ –∑–∞–ø–∏—Å–∏ - –≤–µ—Å –•—ç–º–º–∏–Ω–≥–∞ - –ø–æ–¥—Ç–≤–µ—Ä–∂–¥–∞–µ—Ç—Å—è —Å–ª–µ–¥—É—é—â–µ–π —Ñ–æ—Ä–º—É–ª–æ–π $l(n) = \lambda(n) + \nu(n) -1$  </p>

2. –†–µ–∞–ª–∏–∑—É–µ–º –∞–ª–≥–æ—Ä–∏—Ç–º –ë—Ä–∞—É—ç—Ä–∞ –¥–ª—è –≤—ã—á–∏—Å–ª–µ–Ω–∏—è –ø—Ä–∏–±–ª–∏–∂–µ–Ω–Ω—ã—Ö –∞–¥–¥–∏—Ç–∏–≤–Ω—ã—Ö —Ü–µ–ø–æ—á–µ–∫

```sage
#–ø–µ—Ä–µ–≤–æ–¥ –≤ 2^k —Å.—Å
def shift_num_sys(n, k):
    new_num = []
    divisor = 1 << k
    while n > 0:
        new_num.append(n%divisor)
        n = n//divisor
    new_num.reverse()
    return new_num

#–í–æ–∑–≤—Ä–∞—â–∞–µ—Ç "–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ü–∏—Ñ—Ä –≤ –¥–≤–æ–∏—á–Ω–æ–π —Å–∏—Å—Ç–µ–º–µ —Å—á–∏—Å–ª–µ–Ω–∏—è - 1"
def _lambda(n):
    return len(shift_num_sys(n, 1))-1

#–ú–∏–Ω–∏–º–∞–ª—å–Ω–∞—è –¥–ª—è–Ω–∞ —Ü–µ–ø–æ—á–∫–∏
def chain_length(n):
    return int(_lambda(n) + _lambda(n)/(_lambda(_lambda(n))) + (_lambda(n)*_lambda(_lambda(_lambda(n))))/((_lambda(_lambda(n)))**2))

#k –ø—Ä–∏ –∫–æ—Ç–æ—Ä–æ–º –¥–ª–∏–Ω–∞ –∞–¥–¥–∏—Ç–∏–≤–Ω–æ–π —Ü–µ–ø–æ—á–∫–∏ –±—É–¥–µ—Ç –º–∏–Ω–∏–º–∞–ª—å–Ω–∞
def best_k(n):
    return _lambda(_lambda(n))-2*_lambda(_lambda(_lambda(n)))


def brouwer(n, k):
    br_chain = []
    n_in_k = shift_num_sys(n, k)
    max_digit = max(n_in_k)
    for i in range(1, max_digit+1):
        br_chain.append(i)
    intermediate = 0
    for z in range(len(n_in_k)):
        temp = n_in_k[z]
        intermediate += temp
        if z != 0:
            br_chain.append(intermediate)
        if intermediate == n:
            break
        for j in range(k):
            intermediate *= 2
            br_chain.append(intermediate)
    return sorted(set(br_chain))


def chain_size(_chain):
    return len(_chain)-1


test_set = [31415, 2000, 5264, 16971, 953]
for val in test_set:
    print("TESTING BRAUWER'S ALGORITHM ON VALUE ", val)
    bk = best_k(val)
    min_chain = brouwer(val, bk)
    print("best_k = ", bk, "\t CHAIN: ", *min_chain)
    print("l(n) = ", chain_size(min_chain))
    for k in range(1, 7):
        chain = brouwer(val, k)
        print("k = ", k, "\t CHAIN: ", *chain)
        print("l(n) = ", chain_size(chain),  "\tMinimal length = ", chain_length(val))
    print("\n")
```

–ú–æ–∂–Ω–æ –∑–∞–º–µ—Ç–∏—Ç—å, —á—Ç–æ –ø–æ –º–µ—Ä–µ —É–≤–µ–ª–∏—á–µ–Ω–∏—è $k$ –¥–ª–∏–Ω–∞ —Ü–µ–ø–æ—á–∫–∏ –∑–Ω–∞—á–∏—Ç–µ–ª—å–Ω–æ —É–≤–µ–ª–∏—á–∏–≤–∞–µ—Ç—Å—è –∏ —Å—Ç–∞–Ω–æ–≤–∏—Ç—Å—è –∑–Ω–∞—á–∏—Ç–µ–ª—å–Ω–æ –±–æ–ª—å—à–µ –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–π –¥–ª–∏–Ω—ã. –î–ª—è —Ç–æ–≥–æ —á—Ç–æ–±—ã –ø–æ–ª—É—á–∞—Ç—å –æ–ø—Ç–∏–º–∞–ª—å–Ω—É—é –¥–ª–∏–Ω—É –∞–¥–¥–∏—Ç–∏–≤–Ω–æ–π —Ü–µ–ø–æ—á–∫–∏ $k$ –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å —Å–æ–≤—Å–µ–º –Ω–µ–≤–µ–ª–∏–∫–æ –ø–æ —Å—Ä–∞–≤–Ω–µ–Ω–∏—é —Å $n$. –¢–∞–∫–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –ø–æ–ª—É—á–∞–µ—Ç—Å—è –ø—Ä–∏ –ø—Ä–∏–º–µ–Ω–µ–Ω–∏–∏ —Ñ–æ—Ä–º—É–ª—ã $k = \lambda\lambda(n)-2\lambda\lambda\lambda(n) $, –¥–ª—è –Ω–µ–±–æ–ª—å—à–∏—Ö —á–∏—Å–µ–ª $k$ —Å–ª–µ–¥—É–µ—Ç –≤–∑—è—Ç—å —Ä–∞–≤–Ω—ã–º 1 –∏–ª–∏ 2.

3. –†–µ–∞–ª–∏–∑—É–µ–º –∞–ª–≥–æ—Ä–∏—Ç–º –¥—Ä–æ–±–ª–µ–Ω–∏—è –≤–µ–∫—Ç–æ—Ä–∞ –∏–Ω–¥–µ–∫—Å–æ–≤ –¥–ª—è –Ω–∞—Ö–æ–∂–¥–µ–Ω–∏—è –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–π –∑–≤—ë–∑–¥–Ω–æ–π —Ü–µ–ø–æ—á–∫–∏ –¥–ª—è –∑–∞–¥–∞–Ω–Ω–æ–≥–æ —á–∏—Å–ª–∞.

```sage
import time
from math import log, ceil, floor

def _lambda(n):
    return len(shift_num_sys(n, 1))-1

def make_chain(vec):
    chain = [1]
    for i in vec:
        chain.append(chain[-1] + chain[i-1])
    return chain


def dec_vector(vector, start, end):
    # —É—Å—Ç–∞–Ω–∞–≤–ª–∏–≤–∞–µ–º –Ω–∞—á–∞–ª—å–Ω—ã–π –∏–Ω–¥–µ–∫—Å –Ω–∞ –∫–æ–Ω–µ—Ü
    i = end
    # –ø—Ä–æ–≤–µ—Ä—è–µ–º —è–≤–ª—è–µ—Ç—Å—è –ª–∏ —Ü–∏—Ñ—Ä–∞ –Ω–∞ –¥–∞–Ω–Ω–æ–π –ø–æ–∑–∏—Ü–∏–∏ 1, –µ—Å–ª–∏ –¥–∞, —Ç–æ —Å–º–µ—â–∞–µ–º—Å—è –Ω–∞ –æ–¥–Ω—É –ø–æ–∑–∏—Ü–∏—é –≤–ª–µ–≤–æ
    while vector[i] == 1 and i >= start:
        i -= 1
    # –µ—Å–ª–∏ –≤—Å–µ —ç–ª–µ–º–µ–Ω—Ç—ã –æ—Ç –Ω–∞—á–∞–ª–∞ –¥–æ –∫–æ–Ω—Ü–∞ —Ä–∞–≤–Ω–æ 1, —Ç–æ —É–º–µ–Ω—å—à–µ–Ω–∏–µ –≤–µ–∫—Ç–æ—Ä–∞ –Ω–µ–≤–æ–∑–º–æ–∂–Ω–æ
    if i < start:
        return False
    # —É–º–µ–Ω—å—à–∞–µ–º –ø–µ—Ä–≤—ã–π, –±–æ–ª—å—à–∏–π 1, —ç–ª–µ–º–µ–Ω—Ç
    vector[i] -= 1
    # –≤—ã—Å—Ç–∞–≤–∏—Ç—å –≤—Å–µ —Å–ª–µ–¥—É—é—â–∏–µ –∏–Ω–¥–µ–∫—Å—ã –≤ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ
    for j in range(i+1, end+1):
        vector[j] = j+1
    return True


def fraction(n):
    if n < 3:
        return list(range(1, n + 1))
    _l = ceil(log(n, 2))
    new_n = shift_num_sys(n, 1)
    nu = new_n.count(1)
    l_ = floor(log(n, 2))
    for m in range(_l, l_+ nu +1):
        vector = list(range(1, m+1))
        q = m//2
        vector[q-1] += 1
        while dec_vector(vector, 0, q-1):
            a = make_chain(vector)
            if a[m] == n:
                return a
            elif (n >= a[q]+m-q) and (n <= a[q]*2**(m-q)):
                while dec_vector(vector, q, len(vector)-1):
                    a = make_chain(vector)
                    if a[m] == n:
                        return a
                for i in range(q, len(vector)):
                    vector[i] = i+1

                    
test_set = [1033, 1040, 2176, 2000, 1234]
for val in test_set:
    print("TESTING FRACTION ON VALUE ", val)
    start = time.time()
    chain = fraction(val)
    end = time.time()
    print("CHAIN: ", *chain)
    print("l(n) = ", chain_size(chain),  "Time elapsed = ", round(end-start, 5), "sec\tMinimal length = ", ceil(log(val, 2)))
    print("\n")                
```

–ê–ª–≥–æ—Ä–∏—Ç–º –¥—Ä–æ–±–ª–µ–Ω–∏—è –≤–µ–∫—Ç–æ—Ä–∞ –∏–Ω–¥–µ–∫—Å–æ–≤ –±—ã–ª –ø—Ä–æ—Ç–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω –Ω–∞ –Ω–∞–±–æ—Ä–µ [1033, 1040, 2176, 2000, 1234]. –ú–∏–Ω–∏–º–∞–ª—å–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤—Ä–µ–º–µ–Ω–∏ –ø–æ—Ç—Ä–µ–±–æ–≤–∞–ª–æ—Å—å –¥–ª—è –≤—ã—á–∏—Å–ª–µ–Ω–∏—è —á–∏—Å–µ–ª $1040 = 2^{10} + 2^4$ –∏ $2176 = 2^{11} + 2^7$, –∞ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ –¥–ª—è —á–∏—Å–ª–∞ 1234.  –¢–∞–∫–æ–µ –¥–æ–ª–≥–æ–µ –≤—Ä–µ–º—è —Ä–∞–±–æ—Ç—ã –æ–±—ä—è—Å–Ω—è–µ—Ç—Å—è —Ç–µ–º, —á—Ç–æ –≤ –∞–ª–≥–æ—Ä–∏—Ç–º–µ –ø—Ä–æ–∏—Å—Ö–æ–¥–∏—Ç –ø–µ—Ä–µ–±–æ—Ä —Ä–∞–∑–ª–∏—á–Ω—ã—Ö –≤–µ–∫—Ç–æ—Ä–æ–≤ –∏–Ω–¥–µ–∫—Å–æ–≤, –æ—Ç—á–µ–≥–æ –∞–ª–≥–æ—Ä–∏—Ç–º –∏–º–µ–µ—Ç —Ñ–∞–∫—Ç–æ—Ä–∏–∞–ª—å–Ω—É—é —Å–ª–æ–∂–Ω–æ—Å—Ç—å: $O( \sum(m-1)!)$.

4. –ü—Ä–æ–≤–µ—Ä–∏–º –≥–∏–ø–æ—Ç–µ–∑—É –®–æ–ª—å—Ü–∞-–ë—Ä–∞—É—ç—Ä–∞ –¥–ª—è —á–∏—Å–µ–ª $ 1 \leq n \leq 12$ —Å –ø–æ–º–æ—â—å—é –∞–ª–≥–æ—Ä–∏—Ç–º–∞ –¥—Ä–æ–±–ª–µ–Ω–∏—è –≤–µ–∫—Ç–æ—Ä–∞ –∏–Ω–¥–µ–∫—Å–æ–≤.

–ì–∏–ø–æ—Ç–µ–∑–∞ –®–æ–ª—å—Ü–∞-–ë—Ä–∞—É—ç—Ä–∞: $$ l(2^n -1) \leq l(n) + n -1$$

–ö–æ–¥ –¥–ª—è –ø—Ä–æ–≤–µ—Ä–∫–∏:

```sage
for i in range (1, 13):
    chain_2n = fraction(2**i-1)
    chain_n = fraction(i)
    print("n = ", i, end='')
    if len(chain_2n)-1 <= len(chain_n)+i-2:
        print("\tTrue\t", len(chain_2n)-1, '<=', len(chain_n)+i-2)
    else:
        print("\tFalse")
        print(chain_n)
        break
```
