## Основные теоретические положения

Множество G с бинарной операцией $\cdot$ называется группой, если:

1. Операция $\cdot$ ассоциативна, то есть (a $\cdot$ b) $\cdot$ c = a $\cdot$ (b $\cdot$ c)
2. Существует нейтральный элемент e, такой что a $\cdot$ e = e $\cdot$ a = a
3. Существует обратный элемент a', такой что a' $\cdot$ a = e

Группа G абелева, если операция $\cdot$ коммутативна

**Def:** 
Группа, порождённая одним элементом называется циклической группой и обозначается $<a>$.
    
**Def:** 
Пусть M = {1, 2, ..., n} конечное множество из n элементов. Группа $S_{n} = S(M)$ всех заимно однозначных отображений множества M на себя с операцией умножения(композиции) называется симметрической группой степени n.

Элементы симметрической группы называются подстановками 
    #S = n!
    
Действие на множестве подстановки $\sigma$ удобно описывать указывая образы всех элементов множества M:

\begin{equation}
\sigma =  
  \begin{pmatrix}
    1 & 2 & ... & n\\
    i_{1} & i_{2} & ... & i_{n} 
  \end{pmatrix}
  , \ где \ i_{n} \neq \sigma(i), k=1, ..., n
\end{equation}

Поскольку подстановки представляют собой взаимно однозначные отображения, то в строке образов все элементы $i_{k}$ различны и исчерпывают всё множество M

Подстановки $\sigma$, $\tau \in S_{n}$ умножаются в соответствии с общим правилом умножения композиции отображений:

$\sigma\tau(i) = \sigma(\tau(i))$

Нейтральный элемент является тождественным подстановке $e(i) = i, i=1, ..., n$

Подстановка называется циклом длины k, если на некотором подмножестве $\{a_{i1}, a_{i2}, ..., a_{ik}\}$ множества M она действует следующим образом: 

\begin{equation}
  \begin{pmatrix}
    a_{i1} & ... & a_{ik-1} & a_{ik} \\
    a_{i2} & ... & a_{ik} & a_{i1} 
  \end{pmatrix}
\end{equation}

а остальные элементы множества M не изменяет.

Два цикла называются независимыми, если они действуют на непересекающихся подмножествах M.

Пусть $\sigma$ - цикл длины k, тогда он пораждает циклическую подгруппу порядка k 

<$\sigma$> = {e, $\sigma$, $\sigma^2$, ..., $\sigma^{k-1}$}

$(\sigma^{S})^{-1} = \sigma^{k-s}$, где S = 1, 2, ..., k-1

**Теорема**

Каждая подстановка $\sigma \neq e$ в $S_{n}$ является произведением независимых циклов. Это размножение в произведение определено однозначно с точностью до порядка следования циклов.
    
**Следствие**

Порядок подстановки $\sigma$ равен LCM(НОК) длин независимых циклов, входящих в разложение $\sigma$. Порядок подстановкиявляется порядком циклическиой подгруппы $<\sigma>$

**Def**:
Пусть G - группа с нейтральным элементом e, а X - множество. Будем говорить, что G действует на X, если задана $G*X -> G$ обладающего для любого $x \in X$ и $g,h \in G$ следующим свойством:

1)ex=x

2)gh(x)=g(h(x))

**Def**: 
Орбитой элемента $x \in X$ под действием G называется множество GX = {gx: $g \in G$}. Число #(Gx) называется длиной орбиты.

**Def**:Неподвижными точками элемента $g \in G$ называют точки $x \in X$ для которых gx = x. Множество неподвижных точек элемента g обозначается G*

**Лемма Бёрнсайда (лемма Коши-Фробениуса)**

Количество орбит действия группы G на множество X равно $N = \dfrac{1}{\#G}\sum_{g \in G}\#x^{g}$



Пусть I - произвольное множество, а C - множество цветов.

Раскраской множества I называются функции из I в C. Множество всех раскрасок обозначается $C^{I}$. Количество всех раскрасок равно $(\#C)^{\#I}$.

Количество раскрасок из $C^{I}$, которые сохраняет данный элемент $g \in G$ равно $(\#C)^{C(\phi(g))}$. Здесь $G(\phi)$ обозначает количество независимых циклов в записи перестановки $\sigma$.

### 1. Аналитическое решение задачи.

Введём обозначения:

$e$ - тождественное преобразование.

$\sigma_{i}$ - поворот тэтраэдра, при котором его вершина становится верхей.

$\tau_{ij}$ - поворот на $\frac{2\pi}{3}$ по часовой стрелке относительно оси, проходящей через вершину.

$\{a_{1}, a_{2}, a_{3}, a_{4}\}$ - множество вершин тэтраэдра.

Количество всех раскрасок: $5^4 = 625$ 

Представим нейтральный элемент в виде перестановки: 
\begin{equation}
e = \bigg(
  \begin{matrix}
    1 & 2 & 3 & 4\\
    1 & 2 & 3 & 4 
  \end{matrix}
\bigg) = (1)(2)(3)(4)  
\end{equation}

Можно представить повороты $\sigma_{i}$ в виде перестановок:

Когда верхней вершиной является $a_{2}$ \begin{equation}
\sigma_{1} = \bigg(
  \begin{matrix}
    1 & 2 & 3 & 4\\
    2 & 3 & 1 & 4 
  \end{matrix}
\bigg) = (1 2 3)(4) - 2 \ цикла  
\end{equation}

Когда верхней вершиной является $a_{3}$ \begin{equation}
\sigma_{2} = \bigg(
  \begin{matrix}
    1 & 2 & 3 & 4\\
    3 & 4 & 1 & 2 
  \end{matrix}
\bigg) = (1 3)(2 4) - 2 \ цикла  
\end{equation}

Когда верхней вершиной является $a_{4}$ \begin{equation}
\sigma_{3} = \bigg(
  \begin{matrix}
    1 & 2 & 3 & 4\\
    4 & 2 & 1 & 3 
  \end{matrix}
\bigg) = (1 4 3)(2) - 2 \ цикла  
\end{equation}

Теперь зафиксируем одну вершину и начнём вращать тэтраэдр на $\frac{2\pi}{3}$. Тогда получим следующие перестановки:
\begin{equation}
\tau e = \left( \begin{array}{ccc}
1 & 2 & 3 & 4\\
1 & 4 & 2 & 3\end{array} \right) = (1)(2 \ 4 \ 3) - 2 \ цикла\qquad
\tau_{2} e = \left( \begin{array}{ccc}
1 & 2 & 3 & 4\\
1 & 3 & 4 & 2\end{array} \right) = (1)(2 \ 3 \ 4) - 2 \ цикла
\end{equation}

\begin{equation}
\tau \sigma = \left( \begin{array}{ccc}
1 & 2 & 3 & 4\\
2 & 4 & 3 & 1\end{array} \right) = (1 \ 2 \ 4)(3) - 2 \ цикла\qquad
\tau_{2} \sigma = \left( \begin{array}{ccc}
1 & 2 & 3 & 4\\
2 & 1 & 4 & 3\end{array} \right) = (1 \ 2)(3 \ 4) - 2 \ цикла
\end{equation}

\begin{equation}
\tau \sigma_{2} = \left( \begin{array}{ccc}
1 & 2 & 3 & 4\\
3 & 2 & 4 & 1\end{array} \right) = (1 \ 3 \ 4)(2) - 2 \ цикла\qquad
\tau_{2} \sigma_{2} = \left( \begin{array}{ccc}
1 & 2 & 3 & 4\\
3 & 1 & 2 & 4\end{array} \right) = (1 \ 3 \ 2)(4) - 2 \ цикла
\end{equation}

\begin{equation}
\tau \sigma_{3} = \left( \begin{array}{ccc}
1 & 2 & 3 & 4\\
4 & 3 & 2 & 1\end{array} \right) = (1 \ 4)(2 \ 3) - 2 \ цикла\qquad
\tau_{2} \sigma_{3} = \left( \begin{array}{ccc}
1 & 2 & 3 & 4\\
4 & 1 & 3 & 2\end{array} \right) = (1 \ 4 \ 2)(3) - 2 \ цикла
\end{equation}

Всего получаем 12 перестановок, где 11 получены поворотом тэтраэдра и 1 нейтральный элемент. Можно заметить, что у всех перестановок (кроме нейтральной) 2 цикла, следовательно: $\#X^{e} = 5^{4}$, $\#X^{\tau_{i}\sigma_{j}} = 5^{2}$ и $\#G = 12$

Воспользуемся леммой Бёрнсайда, чтобы узнать кол-во раскрасок:
$$N=\frac{5^{4}+11 \cdot 5^{2}}{12} = 75$$

Для проверки вычислений была составлена функция, составляющая все перестановки:

```python
sigma = "(1)(2, 4, 3)" # поворот на 60 градусов по часовой стрелке
tau = "(1, 2)(3, 4)" # переворот на лицевую сторону
G = PermutationGroup([sigma, tau]) # всевозможные комбинации sigma, tau
print('Пусть G - множество всех комбинаций поворотов тэтраэдра. Все элементы G:')
for i in range(G.order()):
     print(i+1, ' -> ', G[i].cycle_string(singletons=True))
```

### 2. Графическое отображение раскраски по заданному номеру.

Реализуем функцию, которая будет отображать раскраску по заданному номеру и по заданным цветам.

```python
def color(num, colors=['red', 'orange', 'blue', 'green', 'black']):
        v_i = [(2, 2, 4),(0, 0, 0), (0, 4, 0), (4, 2, 0)] #координаты вершин тэтраэдра
        lines = [(v_i[0], v_i[1]), (v_i[1], v_i[2]), (v_i[2], v_i[3]), (v_i[0], v_i[3]), (v_i[1], v_i[3]), (v_i[0], v_i[2])]#отрезки, соединяющие вершины
        G = Graphics()
        for i in lines:
            G += line3d(i, color='cyan', thickness=0.3)
        for i in range(4):
            G += point3d(v_i[i], color=colors[num%len(colors)], size=30)
            G += text3d(str(i+1), (v_i[i][0], v_i[i][1], v_i[i][2]+0.2), color=colors[num%len(colors)], fontsize=20)
            num //=  len(colors)
        return G
```

$$N=\frac{k^{4}+11 \cdot k^{2}}{12}$$

$k=3 \ \  N=\frac{3^{4}+11 \cdot 3^{2}}{12} = 15$
