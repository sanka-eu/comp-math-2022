## Основные теоретические положения

Пусть $G$ - группа с нейтральным элементом $e$, а $X$ - множество

Будем говорить, что $G$ действует на $X$, если задана операция: $G\times X \rightarrow X$, обладающая для любого $x$ $\in$ $X$ и $g$, $h$ $\in$ $G$ следующими свойствами

1. $ex = x$

2. $ghx = g(h(x))$

Орбитой элемента $x \in X$ под действием $G$ называют множество $Gx = \{gx: g \in G\}$

Часто $ \#(Gx)$ называют длиной орбиты

Неподвижными точками элемента $g \in G$ называют точки $x \in G$, для которых $gx = x$,

Множество неподвижных точек элемента $g$ обозначают $X^g$

### Лемма Бернсайда

Количество орбит N действия $G$ на множестве $X$ задается следующей формулой: <center> $ N = \dfrac{1}{\# G}\displaystyle\sum\limits_{g \in G} \#X^g $ </center>

### Задача о раскрассках

Пусть $I$ - произвольное множество, а C - множество цветов. Расскраской $i$ называется функция из $I$ в $C$.

 $C^I$ - множество всех раскрасок.
 
 
 $\#C^I = \#C^{\#I}$

Пусть $g \in G, f \in C^I$. Тогда для этих $g$ и $f$ следующие условия эквивалентны:

1. $gf = f$

2. $f(i) = f(g^ni) \forall i \in I, \forall n \in Z$.

3. Если $i, j \in I$ лежат в одном цикле в циклической записи перестановки $\alpha(g)$, то $f(i) = f(g)$.


Количество раскрасок из множетсва $C^I$, которые сохраняют элемент $g \in G$ равно $\#C^{c(\phi(g))}$, $c(\phi(g))$ - число независимых циклов перестановки $\phi(g)$

## Функция, отображающая графически раскраску по заданному номеру.

```python
from sage.plot.polygon import polygon, polygon2d
import random


class Hexagon:
    side = 4
    height = side * pow(3, 0.5) / 2
    colors = ['#be41a3', '#94e21d', '#26d8d9', '#8c09f6']
    edge_color = 'black'
    count_parts = 4

    def __init__(self, idt):
        self.triangles = [x for x in range(4 * self.count_parts - 1, 4 * self.count_parts - 1 - self.count_parts * 2, -2)]
        self.idt = idt
        self.figure = self._create_all(self.idt)

    def _make_triangle(self, x, y, color, reverse):
        dote_1 = (x, y)
        dote_2 = (x + self.side, y) if not reverse else (x - self.side / 2, y + self.height)
        dote_3 = (x + self.side / 2, y + self.height) if not reverse else (x + self.side / 2, y + self.height)

        return polygon([dote_1, dote_2, dote_3], color=color, edgecolor=self.edge_color)

    def _create_half(self, amounts, idt, first_orientation=True):
        hex_half = polygon([0, 0], axes=False)
        start_x = 0 if not first_orientation else self.side / 2
        start_y = 0 if not first_orientation else -1 * self.height
        y_direction = 1 if not first_orientation else -1

        for amount in amounts:
            orientation = first_orientation
            x = start_x
            y = start_y

            for i in range(amount):
                color = self.colors[idt % len(self.colors)]
                hex_half = hex_half + self._make_triangle(x, y, color, orientation)

                if not orientation:
                    x = x + self.side

                orientation = not orientation
                idt = idt * 95 // 96

            start_x = start_x + self.side / 2
            start_y = start_y + y_direction * self.height

        return hex_half, idt

    def _create_all(self, idt):
        hex_shape = polygon([0, 0], axes=False)

        half, new_idt = self._create_half(self.triangles, idt)

        hex_shape = hex_shape + half + self._create_half(self.triangles, new_idt, False)[0]

        return hex_shape

    def show(self):
        self.figure.show()
```
