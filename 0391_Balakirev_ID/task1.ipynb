### –ê–ª–≥–æ—Ä–∏—Ç–º –ë—Ä–∞—É—ç—Ä–∞:

–î–ª—è –∑–∞–¥–∞–Ω–Ω–æ–≥–æ —á–∏—Å–ª–∞ $ n \in \mathbb{N} $ –º–æ–∂–Ω–æ –ø–æ—Å—Ç—Ä–æ–∏—Ç—å —Ü–µ–ø–æ—á–∫—É –ë—Ä–∞—É—ç—Ä–∞ —Å –ø–æ–º–æ—â—å—é —Ä–µ–∫—É—Ä—Ä–µ–Ω—Ç–Ω–æ–π —Ñ–æ—Ä–º—É–ª—ã

$$ B_k(n) = 
\begin{cases}

–î–ª–∏–Ω–∞ —Ü–µ–ø–æ—á–∫–∏ –±—É–¥–µ—Ç –º–∏–Ω–∏–º–∏–∑–∏—Ä–æ–≤–∞–Ω–∞ –¥–ª—è –±–æ–ª—å—à–∏—Ö $ n $, –µ—Å–ª–∏ 
$$ k = \lambda(\lambda(n)) - 2\lambda(\lambda(\lambda(n))) $$

### –£—Ç–æ—á–Ω–µ–Ω–∏—è:

* –ó–∞–¥–∞–µ—Ç—Å—è —Ñ–∏–∫–∏—Ä–æ–≤–∞–Ω–Ω—ã–π –ø–∞—Ä–∞–º–µ—Ç—Ä $ k $. –í—ã–ø–æ–ª–Ω—è—é—Ç—Å—è –≤—ã—á–∏—Å–ª–µ–Ω–∏—è –≤—Å–ø–æ–º–æ–≥–∞—Ç–µ–ª—å–Ω—ã—Ö —á–∏—Å–µ–ª:

$$

q_2 = [\frac{q_1}{d}], r_2 = q_1 \ mod \ d \Rightarrow q_1 = q_{2}d + r_{2}
$$

* –î–∞–Ω–Ω–∞—è –ø—Ä–æ—Ü–µ–¥—É—Ä–∞ –≤—ã–ø–æ–ª–Ω—è–µ—Ç—Å—è –¥–æ —Ç–µ—Ö –ø–æ—Ä, –ø–æ–∫–∞ $ q_s $ –Ω–µ —Å—Ç–∞–Ω–µ—Ç –º–µ–Ω—å—à–µ $ d $. –°–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ $ q_{s-1} = q_{s}d + r_{s} $

* –¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º,
$$ n = 2^kq_1 + r_1 = ... = 2^k(2^k(...(2^kq_s + r_s)...) + r_2) + r_1 $$
$$ B_k(n): 1, 2, 3, ..., 2^k - 1, 2q_s, 4q_s, ..., 2^kq_s + r_s, ..., n $$

### –ó–∞–¥–∞–Ω–∏–µ 2

–†–µ–∞–ª–∏–∑–æ–≤–∞—Ç—å –∞–ª–≥–æ—Ä–∏—Ç–º –ë—Ä–∞—É—ç—Ä–∞ –¥–ª—è –≤—ã—á–∏—Å–ª–µ–Ω–∏—è –ø—Ä–∏–±–ª–∏–∂—ë–Ω–Ω—ã—Ö –∞–¥–¥–∏—Ç–∏–≤–Ω—ã—Ö —Ü–µ–ø–æ—á–µ–∫ –¥–ª—è —Ä–∞–∑–ª–∏—á–Ω—ã—Ö —á–∏—Å–µ–ª –ø—Ä–∏ –≤–∞—Ä—å–∏—Ä–æ–≤–∞–Ω–∏–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ k, —Å–æ–ø–æ—Å—Ç–∞–≤–∏—Ç—å –¥–ª–∏–Ω—ã –ø–æ–ª—É—á–µ–Ω–Ω—ã—Ö –∞–¥–¥–∏—Ç–∏–≤–Ω—ã—Ö —Ü–µ–ø–æ—á–µ–∫ —Å –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–π –∞–¥–¥–∏—Ç–∏–≤–Ω–æ–π —Ü–µ–ø–æ—á–∫–æ–π –¥–ª—è –∑–∞–¥–∞–Ω–Ω–æ–≥–æ —á–∏—Å–ª–∞. –°–¥–µ–ª–∞—Ç—å –≤—ã–≤–æ–¥—ã.

```sage
def _lambda(n):
	n = Integer(n)
	return n.nbits() - 1


def min_chain_length(n):
	return int(_lambda(n) + _lambda(n) / _lambda(_lambda(n)) + (_lambda(n) * _lambda(_lambda(_lambda(n)))) / (_lambda(_lambda(n)) ** 2))


def chain_remove_dups(chain):
	ret = list(dict.fromkeys(chain))
	if ret[-1] == 0:
		del ret[-1]
	return ret


def chain_to_str(chain):
	return ' '.join(map(str, chain[:min(len(chain), 17)])) + ('...' if len(chain) > 17 else '')


def brouwer(n, k):
	if type(n) not in [int, Integer]:
		raise(TypeError('n must be an integer'))
	if type(k) not in [int, Integer]:
		raise(TypeError('k must be an integer'))
	n = Integer(n)

	d = 2 ** k
	q = n.quo_rem(d)[0]
	rarr = [n.quo_rem(d)[1]]
	sqmax = rarr[0]  # –ú–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ –≤ —Ü–∏—Ñ—Ä–∞—Ö n

	i = 0
	while q >= d:
		rarr.append(q.quo_rem(d)[1])
		q = q.quo_rem(d)[0]
		if rarr[i] > sqmax:
			sqmax = rarr[i]
		i += 1
	rarr.append(q)
	sqmax = max(q, sqmax)

	chain = [0] * (sqmax + (len(rarr) - 1) * (k + 1))  # –¶–µ–ø–æ—á–∫–∞ —Å–ª–æ–∂–µ–Ω–∏—è
	for i in range(1, sqmax + 1):
		# –ù–∞—á–∞–ª—å–Ω–∞—è –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å 1, 2, 3, ..., b, –≥–¥–µ b - –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ –≤ —Ü–∏—Ñ—Ä–∞—Ö n
		chain[i-1] = i

	chain[sqmax - 1] = q
	for i in range(len(rarr) - 1):  # i - –ù–æ–º–µ—Ä —á–∏—Å–ª–∞ –≤ —Ü–µ–ø–æ—á–∫–µ
		for j in range(k):  # j - –ß–∏—Å–ª–æ –¥–ª—è —É–¥–≤–æ–µ–Ω–∏—è
			chain[sqmax + i * (k + 1) + j] = chain[sqmax + i * (k + 1) + j - 1] * 2
		chain[sqmax + i * (k + 1) + k] = chain[sqmax + i *
		(k + 1) + k - 1] + rarr[len(rarr) - 2 - i]
	chain[sqmax - 1] = sqmax

	return chain_remove_dups(chain)


n_set = [randint(17, 127) for _ in range(5)]
n_set.extend(randint(127, 65535) for _ in range(5))
for k in [1, 2, 3, 6, 9, 16]:
	print(f'k = {k}:')
	for n in n_set:
		chain = brouwer(n, k)
		print(f'\tn = {n}\n\tl_B(n) = {len(chain)}\n\tl(n) = {min_chain_length(n)} | {chain_to_str(chain)}\n')

```

__–í—ã–≤–æ–¥:__
–í–∏–¥–Ω–æ, —á—Ç–æ –ø—Ä–∏ –Ω–µ–∫–æ—Ç–æ—Ä—ã—Ö k –∞–ª–≥–æ—Ä–∏—Ç–º –ë—Ä–∞—É—ç—Ä–∞ –¥–∞—ë—Ç –æ—á–µ–Ω—å –±–ª–∏–∑–∫–∏–π –∫ –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–º—É —Ä–µ–∑—É–ª—å—Ç–∞—Ç, –ø—Ä–∏ –¥—Ä—É–≥–∏—Ö k - –∑–Ω–∞—á–∏—Ç–µ–ª—å–Ω–æ –±–æ–ª–µ–µ –¥–ª–∏–Ω–Ω—ã–π. –û—Ü–µ–Ω–∫–∞ –∞–ª–≥–æ—Ä–∏—Ç–º–∞ –ë—Ä–∞—É—ç—Ä–∞ –≥–æ–≤–æ—Ä–∏—Ç –æ —Ç–æ–º, —á—Ç–æ $$ l(n)\approx\lambda(n)+\frac{\lambda(n)}{\lambda(\lambda(n))}$$


–í —Ç–æ –≤—Ä–µ–º—è –∫–∞–∫ —Å–∞–º–∞—è –∫–æ—Ä–æ—Ç–∫–∞—è –∞–¥–¥–∏—Ç–∏–≤–Ω–∞—è —Ü–µ–ø–æ—á–∫–∞
–¥–ª—è —á–∏—Å–ª–∞ n –∏–º–µ–µ—Ç –¥–ª–∏–Ω—É –Ω–µ –±–æ–ª–µ–µ $$ \lambda(n)$$

–ß—Ç–æ–±—ã –∞–ª–≥–æ—Ä–∏—Ç–º –ë—Ä–∞—É—ç—Ä–∞ —Å—Ç—Ä–æ–∏–ª –∞–¥–¥–∏—Ç–∏–≤–Ω—ã–µ —Ü–µ–ø–æ—á–∫–∏ –∞–¥–µ–∫–≤–∞—Ç–Ω–æ–π –¥–ª–∏–Ω—ã, $k$ –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å –∫—É–¥–∞ –º–µ–Ω—å—à–µ $n$ –∏ –¥–ª—è –¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ –±–æ–ª—å—à–∏—Ö $n$ –≤—ã—á–∏—Å–ª—è—Ç—å—Å—è –ø–æ —Ñ–æ—Ä–º—É–ª–µ

$$ k = \lambda \lambda (n) - 2 \lambda \lambda \lambda (n) $$

–∞ –¥–ª—è –Ω–µ–±–æ–ª—å—à–∏—Ö $n$ –±—ã—Ç—å –ø–æ–ª–æ–∂–µ–Ω–æ —Ä–∞–≤–Ω—ã–º $1$ –∏–ª–∏ $2$ (—Ä–∞–∑–Ω–∏—Ü–∞ –¥–æ–ª–∂–Ω–∞ –±—ã—Ç—å –Ω–µ –æ—á–µ–Ω—å –∑–∞–º–µ—Ç–Ω–æ–π).

### –ó–∞–¥–∞–Ω–∏–µ 3

–†–µ–∞–ª–∏–∑–æ–≤–∞—Ç—å –∞–ª–≥–æ—Ä–∏—Ç–º –¥—Ä–æ–±–ª–µ–Ω–∏—è –≤–µ–∫—Ç–æ—Ä–∞ –∏–Ω–¥–µ–∫—Å–æ–≤ –¥–ª—è –Ω–∞—Ö–æ–∂–¥–µ–Ω–∏—è –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–π –∑–≤—ë–∑–¥–Ω–æ–π —Ü–µ–ø–æ—á–∫–∏ –¥–ª—è –∑–∞–¥–∞–Ω–Ω–æ–≥–æ —á–∏—Å–ª–∞. –ü—Ä–æ—Ç–µ—Å—Ç–∏—Ä–æ–≤–∞—Ç—å –∞–ª–≥–æ—Ä–∏—Ç–º –º–∏–Ω–∏–º—É–º –¥–ª—è 5 –∑–Ω–∞—á–µ–Ω–∏–π ùëõ > 1000. –£–∫–∞–∑–∞—Ç—å,
—Å–∫–æ–ª—å–∫–æ –≤—Ä–µ–º–µ–Ω–∏ –ø–æ—Ç—Ä–µ–±–æ–≤–∞–ª–æ—Å—å –Ω–∞ –ø–æ–∏—Å–∫ —Ü–µ–ø–æ—á–∫–∏ –∏ –∫–∞–∫–∞—è —Ü–µ–ø–æ—á–∫–∞
–ø–æ–ª—É—á–∏–ª–∞—Å—å. –°—Ä–∞–≤–Ω–∏—Ç—å —Å –ø—Ä–µ–¥—ã–¥—É—â–∏–º–∏ –º–µ—Ç–æ–¥–∞–º–∏, —Å–¥–µ–ª–∞—Ç—å –≤—ã–≤–æ–¥—ã.

```sage
def _lambda(n):
	n = Integer(n)
	return n.nbits() - 1


def hamming(n):
	n = Integer(n)
	return n.popcount()


def min_chain_length(n):
	return _lambda(n) + 1


def chain_from_indvec(indvec):
	indvec_len = len(indvec)
	nums = [0] * (indvec_len + 1)
	nums[0] = 1

	for i in range(indvec_len):
		nums[i + 1] = nums[i] + nums[indvec[i] - 1]
	return nums


def chain_to_str(chain):
	return " ".join(map(str, chain[:min(len(chain), 40)])) + (" ..." if len(chain) > 40 else "")


def indvec_gen(n, start=1):
	indvec = [0] * n
	for i in range(n):
		indvec[i] = i + start
	return indvec


def indvec_is_min(indvec):
	return all(i <= 1 for i in indvec)


def indvec_dec(indvec):
	indvec_len = len(indvec)

	for i in range(indvec_len):
		if indvec[indvec_len - 1 - i] > 1:
			indvec[len(indvec) - 1 - i] -= 1
			for j in range(indvec_len - i, indvec_len):
				indvec[j] = j + 1
			return


def indvec_get_last_num(indvec):
	indvec_len = len(indvec)
	nums = [0] * (indvec_len + 1)
	nums[0] = 1

	for i in range(indvec_len):
		nums[i + 1] = nums[i] + nums[indvec[i] - 1]
	return nums[indvec_len]


def indvec_crush(n):
	tries = [1, 1]

	lmin = math.ceil(math.log(n, 2))
	lmax = _lambda(n) + hamming(n) - 1

	for m in range(lmin, lmax + 1):
		q = int(m / 2)
		iv_fix = indvec_gen(q)

		while True:  # –û–±—Ä–∞–±–æ—Ç–∫–∞ –≤—Å–µ—Ö –≤–æ–∑–º–æ–∂–Ω—ã—Ö —Ñ–∏–∫—Å–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö —á–∞—Å—Ç–µ–π
			iv_mut = indvec_gen(m - q, q + 1)

			iv_fix.append(iv_mut[0])
			aq = indvec_get_last_num(iv_fix)
			del iv_fix[-1]
			amin = 1 + m - q 		# a_{q+1} + m - q
			amax = aq * 2 ** (m - q) 	# a_{q+1} * 2^{m-q}

			# n –Ω–µ –Ω–∞—Ö–æ–¥–∏—Ç—Å—è –≤ [a_min, a_max], –ø—Ä–æ–ø—É—Å–∫–∞—è —ç—Ç—É —Ñ–∏–∫—Å–∏—Ä–æ–≤–∞–Ω–Ω—É—é —á–∞—Å—Ç—å
			if n > amax or n < amin:
				if indvec_is_min(iv_fix):
					break
				indvec_dec(iv_fix)
				continue

			while True:  # –û–±—Ä–∞–±–æ—Ç–∫–∞ –≤—Å–µ—Ö –≤–æ–∑–º–æ–∂–Ω—ã—Ö –∏–∑–º–µ–Ω—è–µ–º—ã—Ö —á–∞—Å—Ç–µ–π
				if indvec_get_last_num(iv_fix + iv_mut) == n:
					return (iv_fix + iv_mut, tries)

				if indvec_is_min(iv_mut):
					break
				indvec_dec(iv_mut)
				tries[1] += 1

			if indvec_is_min(iv_fix):
				break
			indvec_dec(iv_fix)
			tries[0] += 1


n_set = [randint(1001, 1250) for _ in range(5)]
print(f'{n_set}\n')
for n in n_set:
	res = indvec_crush(n)
	print(f'\tn = {n}\n\tl*(n) = {len(res[0])}\n\tl(n) = {min_chain_length(n)}\n\tFix. parts tried: {res[1][0]}  \n\tMut. parts tried: {res[1][1]}\n{chain_to_str(chain_from_indvec(res[0]))}\n')
```

__–í—ã–≤–æ–¥:__
–ü–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞–º –≤–∏–¥–Ω–æ, —á—Ç–æ —á–µ–º –±–ª–∏–∂–µ —á–∏—Å–ª–æ –∫ —Å—Ç–µ–ø–µ–Ω–∏ –¥–≤–æ–π–∫–∏ (–Ω–∞—á–∏–Ω–∞—è —Å –º–ª–∞–¥—à–∏—Ö —Ä–∞–∑—Ä—è–¥–æ–≤), —Ç–µ–º –º–µ–Ω—å—à–µ –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –ø–µ—Ä–µ–±—Ä–∞—Ç—å –≤–µ–∫—Ç–æ—Ä–æ–≤ –∏–Ω–¥–µ–∫—Å–æ–≤ (—Ä–∞–±–æ—Ç–∞–µ—Ç –æ–Ω __–æ—á–µ–Ω—å —É–∂ –¥–æ–ª–≥–æ__).

–î–ª–∏–Ω—ã –∑–≤—ë–∑–¥–Ω—ã—Ö —Ü–µ–ø–æ—á–µ–∫ –ø–æ—á—Ç–∏ –≤—Å–µ–≥–¥–∞ —Ç–∞–∫–∏–µ –∂–µ, –∫–∞–∫ –∏ –¥–ª—è –∞–ª–≥–æ—Ä–∏—Ç–º–∞ –ë—Ä–∞—É—ç—Ä–∞.

–ü–æ —Å—Ä–∞–≤–Ω–µ–Ω–∏—é —Å –º–µ—Ç–æ–¥–æ–º –ë—Ä–∞—É—ç—Ä–∞ –æ—Ç–Ω–æ—à–µ–Ω–∏–µ –¥–ª–∏–Ω —Ü–µ–ø–æ—á–µ–∫ –∫ —á–∏—Å–ª–∞–º –ø–æ–ª—É—á–∞—é—Ç—Å—è –ø—Ä–∏–º–µ—Ä–Ω–æ –æ–¥–∏–Ω–∞–∫–æ–≤—ã–º–∏, –∏ —Ä–∞–∑—É–º–µ–µ—Ç—Å—è –±–æ–ª—å—à–∏–º–∏ —á–µ–º –º–∏–Ω–∏–º–∞–ª—å–Ω—ã–µ –¥–ª–∏–Ω—ã —Ü–µ–ø–æ—á–µ–∫ 
$l(n) = \lceil lb(n) \rceil$

–¢–∞–∫–∂–µ –≤–∏–¥–Ω–æ, —á—Ç–æ –∞–ª–≥–æ—Ä–∏—Ç–º –∏–º–µ–µ—Ç —Å—Ä–µ–¥–Ω—é—é —Å–ª–æ–∂–Ω–æ—Å—Ç—å $O(\sum(m-1)!)$, –≥–¥–µ $m$ - –¥–ª–∏–Ω–∞ –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–π –∑–≤—ë–∑–¥–Ω–æ–π —Ü–µ–ø–æ—á–∫–∏.


### –ó–∞–¥–∞–Ω–∏–µ 4

–ü—Ä–æ–≤–µ—Ä–∏—Ç—å –≥–∏–ø–æ—Ç–µ–∑—É –®–æ–ª—å—Ü–∞‚Äì–ë—Ä–∞—É—ç—Ä–∞ –¥–ª—è –≤—Å–µ—Ö –Ω–∞—Ç—É—Ä–∞–ª—å–Ω—ã—Ö —á–∏—Å–µ–ª –æ—Ç 1 –¥–æ 12 –Ω–∞ –∞–ª–≥–æ—Ä–∏—Ç–º–µ –¥—Ä–æ–±–ª–µ–Ω–∏—è –≤–µ–∫—Ç–æ—Ä–∞ –∏–Ω–¥–µ–∫—Å–æ–≤. –†–µ–∑—É–ª—å—Ç–∞—Ç—ã –æ—Ñ–æ—Ä–º–∏—Ç—å –≤ –≤–∏–¥–µ —Ç–∞–±–ª–∏—Ü—ã. –°–¥–µ–ª–∞—Ç—å –≤—ã–≤–æ–¥—ã.

```sage
def _lambda(n):
	n = Integer(n)
	return n.nbits() - 1


def hamming(n):
	n = Integer(n)
	return n.popcount()


def min_chain_length(n):
	return _lambda(n) + 1


def chain_from_indvec(indvec):
	indvec_len = len(indvec)
	nums = [0] * (indvec_len + 1)
	nums[0] = 1

	for i in range(indvec_len):
		nums[i + 1] = nums[i] + nums[indvec[i] - 1]
	return nums


def indvec_gen(n, start=1):
	indvec = [0] * n
	for i in range(n):
		indvec[i] = i + start
	return indvec


def indvec_is_min(indvec):
	return all(i <= 1 for i in indvec)


def indvec_dec(indvec):
	indvec_len = len(indvec)

	for i in range(indvec_len):
		if indvec[indvec_len - 1 - i] > 1:
			indvec[len(indvec) - 1 - i] -= 1
			for j in range(indvec_len - i, indvec_len):
				indvec[j] = j + 1
			return


def indvec_get_last_num(indvec):
	indvec_len = len(indvec)
	nums = [0] * (indvec_len + 1)
	nums[0] = 1

	for i in range(indvec_len):
		nums[i + 1] = nums[i] + nums[indvec[i] - 1]
	return nums[indvec_len]


def indvec_crush(n):
	tries = [1, 1]

	lmin = math.ceil(math.log(n, 2))
	lmax = _lambda(n) + hamming(n) - 1

	for m in range(lmin, lmax + 1):
		q = int(m / 2)
		iv_fix = indvec_gen(q)

		while True:  # –û–±—Ä–∞–±–æ—Ç–∫–∞ –≤—Å–µ—Ö –≤–æ–∑–º–æ–∂–Ω—ã—Ö —Ñ–∏–∫—Å–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö —á–∞—Å—Ç–µ–π
			iv_mut = indvec_gen(m - q, q + 1)

			iv_fix.append(iv_mut[0])
			aq = indvec_get_last_num(iv_fix)
			del iv_fix[-1]
			amin = 1 + m - q 		# a_{q+1} + m - q
			amax = aq * 2 ** (m - q) 	# a_{q+1} * 2^{m-q}

			# n –Ω–µ –Ω–∞—Ö–æ–¥–∏—Ç—Å—è –≤ [a_min, a_max], –ø—Ä–æ–ø—É—Å–∫–∞—è —ç—Ç—É —Ñ–∏–∫—Å–∏—Ä–æ–≤–∞–Ω–Ω—É—é —á–∞—Å—Ç—å
			if n > amax or n < amin:
				if indvec_is_min(iv_fix):
					break
				indvec_dec(iv_fix)
				continue

			while True:  # –û–±—Ä–∞–±–æ—Ç–∫–∞ –≤—Å–µ—Ö –≤–æ–∑–º–æ–∂–Ω—ã—Ö –∏–∑–º–µ–Ω—è–µ–º—ã—Ö —á–∞—Å—Ç–µ–π
				if indvec_get_last_num(iv_fix + iv_mut) == n:
					return (iv_fix + iv_mut, tries)

				if indvec_is_min(iv_mut):
					break
				indvec_dec(iv_mut)
				tries[1] += 1

			if indvec_is_min(iv_fix):
				break
			indvec_dec(iv_fix)
			tries[0] += 1


print('l*(2^n-1) <= l*(n) + n - 1')
for n in range(2, 13):
	res = indvec_crush(n)
	resb = indvec_crush(2 ** n - 1)
	print(f'{len(resb[0])} <= {len(res[0])} + {n} - 1\t -> {len(resb[0]) <= len(res[0]) + n - 1}')
```
