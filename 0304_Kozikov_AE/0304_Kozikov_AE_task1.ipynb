## –¶–µ–ª—å —Ä–∞–±–æ—Ç—ã

–§–æ—Ä–º–∏—Ä–æ–≤–∞–Ω–∏–µ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏—è –æ –∞–¥–¥–∏—Ç–∏–≤–Ω—ã—Ö —Ü–µ–ø–æ—á–∫–∞—Ö, –≤—ã—Ä–∞–±–æ—Ç–∞—Ç—å —É–º–µ–Ω–∏–µ —Å–æ—Å—Ç–∞–≤–ª—è—Ç—å –∏ –ø—Ä–∏–º–µ–Ω—è—Ç—å –∞–ª–≥–æ—Ä–∏—Ç–º—ã –¥–ª—è –Ω–∞—Ö–æ–∂–¥–µ–Ω–∏—è –º–∏–Ω–∏–º–∞–ª—å–Ω—ã—Ö

## –û—Å–Ω–æ–≤–Ω—ã–µ —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∏–µ –ø–æ–ª–æ–∂–µ–Ω–∏—è
–ê–¥–¥–∏—Ç–∏–≤–Ω–∞—è —Ü–µ–ø–æ—á–∫–∞ –¥–ª—è —á–∏—Å–ª–∞ $n \in ‚Ñï$ –Ω–∞–∑—ã–≤–∞–µ—Ç—Å—è –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å –Ω–∞—Ç—É—Ä–∞–ª—å–Ω—ã—Ö —á–∏—Å–µ–ª, —Ç–∞–∫–∏—Ö —á—Ç–æ: 
$$ 1 = a_0, a_1, ..., a_r = n $$
$$ a_i = a_j + a_k, \quad k \le j \le i, \quad i = 1, 2, ..., r $$
–û–±–æ–∑–Ω–∞—á–µ–Ω–∏—è: 

$l(n)$ ‚Äî –¥–ª–∏–Ω–∞ –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–π –∞–¥–¥–∏—Ç–∏–≤–Ω–æ–π —Ü–µ–ø–æ—á–∫–∏ –¥–ª—è $n$.
$\newline \lambda(n) = \lfloor lb(n) \rfloor$
$\nu(n)$ ‚Äî –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –µ–¥–∏–Ω–∏—Ü –≤ –¥–≤–æ–∏—á–Ω–æ–º –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏–∏.

–û—Ü–µ–Ω–∫–∏ –¥–ª–∏–Ω—ã –¥–ª—è –∞–¥–¥–∏—Ç–∏–≤–Ω–æ–π —Ü–µ–ø–æ—á–∫–∏:

–ú–µ—Ç–æ–¥ –º–Ω–æ–∂–∏—Ç–µ–ª–µ–π: $$l(mn) \leq l(m) + l(n) $$
m-–∞—Ä–Ω—ã–π –º–µ—Ç–æ–¥: $$l(n) \leq m - 2 + (k + 1)^t$$
–ë–∏–Ω–∞—Ä–Ω—ã–π –º–µ—Ç–æ–¥ SX: $$l(n) \leq \lambda(n) + \nu(n) - 1$$
### –ê–ª–≥–æ—Ä–∏—Ç–º –Ø–æ
–†–∞—Å—Å–º–æ—Ç—Ä–∏–º –∑–∞–¥–∞—á—É –æ –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏–∏ –∞–¥–¥–∏—Ç–∏–≤–Ω–æ–π —Ü–µ–ø–æ—á–∫–∏ –ø—Ä–∏ —É—Å–ª–æ–≤–∏–∏ $k \leq 2$ –∏ $n \in ‚Ñï$. 

1. –†–∞–∑–ª–æ–∂–∏–º $n$ –≤ —Å–∏—Å—Ç–µ–º–µ —Å—á–∏—Å–ª–µ–Ω–∏—è $2^k$
2. –í–≤–µ–¥—ë–º —Ñ—É–Ω–∫—Ü–∏—é: $d(z) = \sum 2^{ik}$ ‚Äî –≥–¥–µ $i: a_i = z$ 
3. –ó–∞–¥–∞–¥–∏–º –±–∞–∑—É: $1, 2, 4, 8, ..., 2^{\lambda(n)}$
4. –í—ã—á–∏—Å–ª–∏–º –∑–Ω–∞—á–µ–Ω–∏—è $d(z)$ –¥–ª—è $z \in (1, 2, 3, ..., 2^k - 1)$ —Ç–∞–∫–∏—Ö, —á—Ç–æ $d(z) \neq 0$
5. –ü–æ—Å—Ç—Ä–æ–∏–º —Ü–µ–ø–æ—á–∫–∏: $d(z), ..., z \cdot d(z)$ –∏ –¥–æ–±–∞–≤–∏–º –∑–Ω–∞—á–µ–Ω–∏—è –±–µ–∑ –ø–æ–≤—Ç–æ—Ä–æ–Ω–∏–π –≤ –±–∞–∑—É.
6. –ü—Ä–æ–≤–µ—Ä–∫—É –∑–Ω–∞—á–µ–Ω–∏–π –º–æ–∂–Ω–æ —Å–¥–µ–ª–∞—Ç—å —Å –ø–æ–º–æ—â—å—é —É—Å–ª–æ–≤–∏—è: $n = \sum_{z=1}^{2^k-1} z \cdot d(z)$
### –ê–ª–≥–æ—Ä–∏—Ç–º –¥—Ä–æ–±–ª–µ–Ω–∏—è –≤–µ–∫—Ç–æ—Ä–∞ –∏–Ω–¥–µ–∫—Å–æ–≤
–†–∞—Å—Å–º–æ—Ç—Ä–∏–º $n \in ‚Ñï$. –ù–µ–æ–±—Ö–æ–¥–∏–º–æ –Ω–∞–π—Ç–∏ –º–∏–Ω–∏–º–∞–ª—å–Ω—É—é –∞–¥–¥–∏—Ç–∏–≤–Ω—É—é —Ü–µ–ø–æ—á–∫—É –¥–ª—è $n$.
–ü—Ä—è–º–æ–π –∞–ª–≥–æ—Ä–∏—Ç–º —Ä–µ–∞–ª–∏–∑–æ–≤–∞–Ω —Å –ø–æ–º–æ—â—å—é –ø–µ—Ä–µ–±–æ—Ä–∞ –≤–µ–∫—Ç–æ—Ä–æ–≤ –∏–Ω–¥–µ–∫—Å–æ–≤. –ü—Ä–∏ —É–º–µ–Ω—å—à–µ–Ω–∏–∏ –ø–æ—Å–ª–µ–¥–Ω–µ–≥–æ —ç–ª–µ–º–µ–Ω—Ç–∞ –≤–µ–∫—Ç–æ—Ä–∞ —Å $m-1$ –¥–æ $1$ –ø–æ—Å–ª–µ–¥–Ω–∏–π —ç–ª–µ–º–µ–Ω—Ç –∑–≤—ë–∑–¥–Ω–æ–π –º–æ–Ω–æ—Ç–æ–Ω–Ω–æ —É–±—ã–≤–∞–µ—Ç, –∞ –∑–Ω–∞—á–∏—Ç –º–æ–∂–Ω–æ —É–º–µ–Ω—å—à–∞—Ç—å –Ω–µ –≤–µ—Å—å –≤–µ–∫—Ç–æ—Ä, –∞ –ª–∏—à—å –µ–≥–æ —á–∞—Å—Ç—å.

–†–∞—Å—Å–º–æ—Ç—Ä–∏–º: $\{r_1, r_2, ..., r_q\}\bigcup\{p_{q+1}, ..., p_m\}$ ‚Äî –≥–¥–µ $m \in [l_{min}(n),l_{max}(n)]$, –∞ $q = m/2$

–í–Ω–µ—à–Ω–∏–π —Ü–∏–∫–ª –æ—Å–Ω–æ–≤–∞–Ω –Ω–∞ –ø–µ—Ä–µ–±–æ—Ä–µ –≤–æ–∑–º–æ–∂–Ω—ã—Ö $m$ –∏–∑ –ø—Ä–æ–º–µ–∂—É—Ç–∫–∞. –í–Ω—É—Ç—Ä–µ–Ω–Ω–∏–π –ø–µ—Ä–µ–±–∏—Ä–∞–µ—Ç $\{r_1,...,r_q\}$, –Ω–∞ –∫–∞–∂–¥–æ–º —à–∞–≥–µ –≤—ã—á–∏—Å–ª—è—è $a_{min}, a_{max}$:
$$a_{min} = a_{q+1} + m - q$$
$$a_{max} = a_{q+1} \cdot 2^{m-q}$$
–ï—Å–ª–∏ –Ω–∞ –æ–¥–Ω–æ–º –∏–∑ —à–∞–≥–æ–≤, –ø–æ—Å–ª–µ–¥–Ω–∏–π —ç–ª–µ–º–µ–Ω—Ç —Ü–µ–ø–∏ = $n$, —Ç–æ –∞–ª–≥–æ—Ä–∏—Ç–º –∑–∞–≤–µ—Ä—à–∞–µ—Ç—Å—è. –¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º –±—É–¥–µ—Ç –ø–µ—Ä–µ–±–∏—Ä–∞—Ç—å—Å—è —Ü–µ–ø—å —Å–Ω–∞—á–∞–ª–∞ –∏–∑ –≤–Ω—É—Ç—Ä–µ–Ω–Ω–µ–≥–æ —Ü–∏–∫–ª–∞, –µ—Å–ª–∏ —ç—Ç–æ –Ω–µ –ø—Ä–∏–Ω–µ—Å—ë—Ç —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞, —Ç–æ –±—É–¥–µ—Ç –ø—Ä–æ–∏—Å—Ö–æ–¥–∏—Ç—å –∏—Ç–µ—Ä–∞—Ü–∏—è –ø–æ –≤–Ω–µ—à–Ω–µ–º—É. 
### –ì–∏–ø–æ—Ç–µ–∑–∞ –®–æ–ª—å—Ü–∞-–ë—Ä–∞—É—ç—Ä–∞:
–ì–∏–ø–æ—Ç–µ–∑–∞ —É—Ç–≤–µ—Ä–∂–¥–∞–µ—Ç, —á—Ç–æ –º–∏–Ω–∏–º–∞–ª—å–Ω—ã–µ –¥–ª–∏–Ω—ã –∞–¥–¥–∏—Ç–∏–≤–Ω—ã—Ö —Ü–µ–ø–æ—á–µ–∫ –¥–ª—è $n \in ‚Ñï$ –º–æ–∂–Ω–æ –æ—Ü–µ–Ω–∏—Ç—å, –∫–∞–∫: 
$$l(2^n-1) \leq l(n) + n - 1 $$
–ì–∏–ø–æ—Ç–µ–∑–∞ –±—ã–ª–∞ –¥–æ–∫–∞–∑–∞–Ω–∞ –¥–ª—è –∑–≤—ë–∑–¥–Ω—ã—Ö —Ü–µ–ø–æ—á–µ–∫.

## –ü–æ—Å—Ç–∞–Ω–æ–≤–∫–∞ –∑–∞–¥–∞—á–∏

–†–µ–∞–ª–∏–∑–æ–≤–∞—Ç—å —Ç–æ—á–Ω—ã–µ –∏ –ø—Ä–∏–±–ª–∏–∂—ë–Ω–Ω—ã–µ –∞–ª–≥–æ—Ä–∏—Ç–º—ã –Ω–∞—Ö–æ–∂–¥–µ–Ω–∏—è –º–∏–Ω–∏–º–∞–ª—å–Ω—ã—Ö –∞–¥–¥–∏—Ç–∏–≤–Ω—ã—Ö —Ü–µ–ø–æ—á–µ–∫ —Å –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ–º —Å–∏—Å—Ç–µ–º—ã –∫–æ–º–ø—å—é—Ç–µ—Ä–Ω–æ–π –º–∞—Ç–µ–º–∞—Ç–∏–∫–∏ SageMath, –ø—Ä–æ–≤–µ—Å—Ç–∏ –∞–Ω–∞–ª–∏–∑ –∞–ª–≥–æ—Ä–∏—Ç–º–æ–≤. –ü–æ–ª—É—á–µ–Ω–Ω—ã–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã
—Å–æ–¥–µ—Ä–∂–∞—Ç–µ–ª—å–Ω–æ –ø—Ä–æ–∏–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∏—Ä–æ–≤–∞—Ç—å.

## –í—ã–ø–æ–ª–Ω–µ–Ω–∏–µ —Ä–∞–±–æ—Ç—ã
1. –í—Ä—É—á–Ω—É—é (—Ç.–µ. –Ω–µ —Ä–µ–∞–ª–∏–∑–æ–≤—ã–≤–∞—è –∞–ª–≥–æ—Ä–∏—Ç–º –Ω–∞ Sage) –ø–æ—Å—Ç—Ä–æ–∏—Ç—å –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å –≤—ã—á–∏—Å–ª–µ–Ω–∏–π –±–∏–Ω–∞—Ä–Ω—ã–º –º–µ—Ç–æ–¥–æ–º –∏ –º–µ—Ç–æ–¥–æ–º –º–Ω–æ–∂–∏—Ç–µ–ª–µ–π –¥–ª—è ùë• ùëõ –¥–ª—è 2-3 –∑–Ω–∞—á–µ–Ω–∏–π ùëõ (–∑–Ω–∞—á–µ–Ω–∏–µ ùëõ > 30 –≤—ã–±–∏—Ä–∞—é—Ç—Å—è —Å—Ç—É–¥–µ–Ω—Ç–æ–º —Å–∞–º–æ—Å—Ç–æ—è—Ç–µ–ª—å–Ω–æ). –°—Ä–∞–≤–Ω–∏—Ç—å –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –æ–ø–µ—Ä–∞—Ü–∏–π –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –º–µ—Ç–æ–¥–∞, —Å–¥–µ–ª–∞—Ç—å –≤—ã–≤–æ–¥—ã.

–†–∞—Å—Å–º–æ—Ç—Ä–∏–º –∑–Ω–∞—á–µ–Ω–∏—è: ${n = 30, 42, 77}$ 

–ë–∏–Ω–∞—Ä–Ω—ã–π –º–µ—Ç–æ–¥:
   1) $n = 30$

–ú–µ—Ç–æ–¥ –º–Ω–æ–∂–∏—Ç–µ–ª–µ–π:
    
   1) $ n = 30 = 2 \cdot 15$  
   $\newline y = x^2$
   $\newline y^{15} = y(y(y(y^2))^2)^2$
   
   2) $ n = 42 = 2 \cdot 21 $ 
   $\newline y = x^2$
   $\newline y^{21} = y((y(y^2)^2)^2)^2$
   
   3) $ n = 77 = 7 \cdot 11 $
   $\newline y = x^7$
   $\newline y^{11} = y(y((y^2)^2))^2$
   
–ö–∞–∫ –º–æ–∂–Ω–æ –∑–∞–º–µ—Ç–∏—Ç—å –∏–∑ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–Ω—ã—Ö —Ä–∞—Å—á—ë—Ç–æ–≤: –º–µ—Ç–æ–¥ –º–Ω–æ–∂–∏—Ç–µ–ª–µ–π —Ö–æ—Ä–æ—à–æ –ø–æ–¥—Ö–æ–¥–∏—Ç –¥–ª—è —á–∏—Å–µ–ª, –∫–æ—Ç–æ—Ä—ã–µ –≤ —Ä–∞–∑–ª–æ–∂–µ–Ω–∏–∏ –¥–∞—é—Ç –Ω–∞–∏–º–µ–Ω—å—à–µ–µ –ø—Ä–æ—Å—Ç–æ–µ —á–∏—Å–ª–æ > 2. –ï—Å–ª–∏ –æ—Ü–µ–Ω–∏—Ç—å –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –æ–ø–µ—Ä–∞—Ü–∏–π –¥–ª—è –±–∏–Ω–∞—Ä–Ω–æ–≥–æ –º–µ—Ç–æ–¥–∞ –∏ –º–µ—Ç–æ–¥–∞ –º–Ω–æ–∂–∏—Ç–µ–ª–µ–π —Ç–æ –ø–æ–ª—É—á–∞–µ—Ç—Å—è:

   1) $n = 30:$ –ë–∏–Ω–∞—Ä–Ω—ã–π –º–µ—Ç–æ–¥: 6. –ú–µ—Ç–æ–¥ –º–Ω–æ–∂–∏—Ç–µ–ª–µ–π: 7.
   
   2) $n = 42:$ –ë–∏–Ω–∞—Ä–Ω—ã–π –º–µ—Ç–æ–¥: 7. –ú–µ—Ç–æ–¥ –º–Ω–æ–∂–∏—Ç–µ–ª–µ–π: 7.
   
   3) $n = 77:$ –ë–∏–Ω–∞—Ä–Ω—ã–π –º–µ—Ç–æ–¥: 8. –ú–µ—Ç–æ–¥ –º–Ω–æ–∂–∏—Ç–µ–ª–µ–π: 6.

2. –†–µ–∞–ª–∏–∑–æ–≤–∞—Ç—å –∞–ª–≥–æ—Ä–∏—Ç–º –Ø–æ –¥–ª—è –≤—ã—á–∏—Å–ª–µ–Ω–∏—è –ø—Ä–∏–±–ª–∏–∂—ë–Ω–Ω—ã—Ö –∞–¥–¥–∏—Ç–∏–≤–Ω—ã—Ö —Ü–µ–ø–æ—á–µ–∫ –¥–ª—è —Ä–∞–∑–ª–∏—á–Ω—ã—Ö —á–∏—Å–µ–ª –ø—Ä–∏ –≤–∞—Ä—å–∏—Ä–æ–≤–∞–Ω–∏–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ ùëò, —Å–æ–ø–æ—Å—Ç–∞–≤–∏—Ç—å –¥–ª–∏–Ω—ã –ø–æ–ª—É—á–µ–Ω–Ω—ã—Ö –∞–¥–¥–∏—Ç–∏–≤–Ω—ã—Ö —Ü–µ–ø–æ—á–µ–∫ —Å –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–π –∞–¥–¥–∏—Ç–∏–≤–Ω–æ–π —Ü–µ–ø–æ—á–∫–æ–π –¥–ª—è –∑–∞–¥–∞–Ω–Ω–æ–≥–æ —á–∏—Å–ª–∞. –°–¥–µ–ª–∞—Ç—å –≤—ã–≤–æ–¥—ã.


```python
class AlgorithmYao:
    def __init__(self):
        self.base = Sequence([0])
        self.zSeq = self.base
        self.nNewSys = list()
        self.binLog = 0
        self.k = 0
        self.n = 0
        
    def d(self, z):
        answer = 0
        for index in range(len(self.nNewSys)):
            if self.nNewSys[index] == z:
                answer += 2**(index*self.k)
        return answer
    
    def build(self, n, k):
        self.k = k
        self.n = n
        self.binLog = log(n, 2) #–±–∏–Ω–∞—Ä–Ω—ã–π –ª–æ–≥–∞—Ä–∏—Ñ–º
        self.base = Sequence([1], immutable = False) #–±–∞–∑–∞ —Ü–µ–ø–æ—á–∫–∏
        while self.base[-1] < 2**floor(self.binLog): 
            self.base.append(self.base[-1]*2)
        self.base.append(n)
        self.zSeq = Sequence(range(1, 2**k), immutable = False) #–ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å z = {1, 2, 3 ... 2^k - 1}
        self.nNewSys = n.digits(2**k) #N –≤ —Å–∏—Å—Ç–µ–º–µ —Å—á–∏—Å–ª–µ–Ω–∏—è 2^k (–Ω–µ —Ä–∞–∑–≤—ë—Ä–Ω—É—Ç—ã–π —Å–ø–∏—Å–æ–∫)
        steps = dict()
        for z in self.nNewSys:
            if z in self.zSeq:
                self.zSeq.remove(z)
                steps.update({z: self.d(z)})
        maxZList = list()
        for step in steps.keys():
            for mul in range(1, step+1):
                if mul*steps[step] not in self.base:
                    if 2*mul*steps[step] < n:
                        self.base.append(2*mul*steps[step])
                    else:
                        self.base.append(mul*steps[step])
            maxZList.append(step*steps[step])
        while len(maxZList) > 2:
            sum = maxZList.pop() + maxZList.pop()
            self.base.append(sum)
            maxZList.append(sum)
        self.base.sort()
        return self.base
    
    def nLambda(self, n):
        return n.nbits() - 1
    
    def minLen(self):
        return self.nLambda(self.n) + self.nLambda(self.n) / self.nLambda(self.nLambda(self.n)) + (self.nLambda(self.n) * self.nLambda(self.nLambda(self.nLambda(self.n)))) / (self.nLambda(self.nLambda(self.n)) ** 2)
        
    def print(self):
        print(f"{self.n}\t|{self.k}\t|{floor(self.minLen())}\t|{len(self.base)}\t|{self.base}")
        

nList = [19, 42, 839, 36171]
print("n\t|k\t|min len|len\t|chain")
print("--------|-------|-------|-------|--------")
for k in range(2, 5):
    c1 = AlgorithmYao()
    for n in nList:
        c1.build(n, k)
        c1.print()
    print("\n")
```

    3.–†–µ–∞–ª–∏–∑–æ–≤–∞—Ç—å –∞–ª–≥–æ—Ä–∏—Ç–º –¥—Ä–æ–±–ª–µ–Ω–∏—è –≤–µ–∫—Ç–æ—Ä–∞ –∏–Ω–¥–µ–∫—Å–æ–≤ –¥–ª—è –Ω–∞—Ö–æ–∂–¥–µ–Ω–∏—è –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–π –∑–≤—ë–∑–¥–Ω–æ–π —Ü–µ–ø–æ—á–∫–∏ –¥–ª—è –∑–∞–¥–∞–Ω–Ω–æ–≥–æ —á–∏—Å–ª–∞. –ü—Ä–æ—Ç–µ—Å—Ç–∏—Ä–æ–≤–∞—Ç—å –∞–ª–≥–æ—Ä–∏—Ç–º –º–∏–Ω–∏–º—É–º –¥–ª—è 5 –∑–Ω–∞—á–µ–Ω–∏–π ùëõ > 1000. –£–∫–∞–∑–∞—Ç—å, —Å–∫–æ–ª—å–∫–æ –≤—Ä–µ–º–µ–Ω–∏ –ø–æ—Ç—Ä–µ–±–æ–≤–∞–ª–æ—Å—å –Ω–∞ –ø–æ–∏—Å–∫ —Ü–µ–ø–æ—á–∫–∏ –∏ –∫–∞–∫–∞—è —Ü–µ–ø–æ—á–∫–∞ –ø–æ–ª—É—á–∏–ª–∞—Å—å. –°—Ä–∞–≤–Ω–∏—Ç—å —Å –ø—Ä–µ–¥—ã–¥—É—â–∏–º–∏ –º–µ—Ç–æ–¥–∞–º–∏, —Å–¥–µ–ª–∞—Ç—å –≤—ã–≤–æ–¥—ã.

```python
import threading
import time

class VecIndex():
    def __init__(self):
        self.chain = Sequence([0])
        pass
        
    def printChain(self):
        return self.chain
    
    def chainBuild(self, fullSeq):
        chain = Sequence([1], immutable = False)
        for i in fullSeq:
            chain.append(chain[-1] + chain[i - 1])
        return chain
    
    def nextSeq(self, seq, q = 0):
        if (seq == [1 for i in range(len(seq))]):
            return [0 for i in range(len(seq))]
        for i in range(len(seq) - 1, -1, -1):
            if (seq[i] == 1):
                seq[i] = i + 1 + q
                continue
            seq[i] -= 1
            break
        return seq
    
    def initBuild(self, n):
        lenMin = math.ceil(log(n, 2))
        lenMax = n.nbits() - 1 + n.popcount() - 1
        support = threading.Thread(target = self.vecBuild, args=(n, lenMin, ceil((lenMin + lenMax)/2)))
        support.start()
        self.vecBuild(n, ceil((lenMin + lenMax)/2), lenMax)
        return self.chain
    
    def vecBuild(self, n, lenMin, lenMax):
        for m in range(lenMin, lenMax + 1):
            if self.chain != [0]:
                return
            q = m // 2
            if q == 0:
                q = 1
            staticSeq = Sequence([i for i in range(1, q + 1)], immutable = False)
            dynamicSeq = Sequence([q + i + 1 for i in range(m - q)], immutable = False)
            while staticSeq != [0 for i in range(q)]:
                chain = self.chainBuild(staticSeq + dynamicSeq)
                aMin = Integer(chain[q] + m - q)
                aMax = Integer(chain[q] * 2 ** (m - q))
                if n < aMin or n > aMax:
                    staticSeq = self.nextSeq(staticSeq)
                    continue
                if chain[-1] == n:
                    self.chain = chain
                    return
                while dynamicSeq != [0 for i in range(m - q)]:
                    chain = self.chainBuild(staticSeq + dynamicSeq)
                    if (n == chain[-1]):
                        self.chain = chain
                        return
                    dynamicSeq = self.nextSeq(dynamicSeq, q)
                staticSeq = self.nextSeq(staticSeq)
                dynamicSeq = Sequence([q + i + 1 for i in range(m - q)], immutable = False)

print("|n\t|time (in min)\t|chain")
print("|-------|---------------|-------")
for num in [1007, 1024, 1234, 1035, 5000]:
    vec = VecIndex()
    start = time.time()
    vec.initBuild(Integer(num))
    end = time.time()
    procTime = round((end - start) / 60, 4)
    print(f"\n|{num}\t|{procTime}\t\t|{vec.printChain()}")
```
