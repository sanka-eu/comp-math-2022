## Основные теоретические положения

### 1) Группа:

Группа - множество G с заданной бинарной операцией $ \times $, если выполнены:

1. Множество замкнуто относительно $ \times $.
2. $ \times $ - ассоциативна, т.е. $ (a \times b) \times c = a \times (b \times c) \forall a, b, c \in G $.
3. $ \exists e \in G $: $ a \times e = e \times a = a, \forall a \in G $.
4. Для $ \forall a \in G $ $ \exists a^{*} \in G $, что $ a^{*} \times a = a \times a^{*} = e $.

### 2) Симметрическая группа:

Пусть $ M = \{ 1, 2, ..., n\} $ - конечное множество из n элементов.

Тогда множество $ S_n = S(M) $ всех взаимно однозначных отображений множества M на себя с операцией умножения (композиции) называется симметрической группой степени n.

$ \cdot $ Элементы симметрической группы называются *подстановками*.

$ \cdot $ Количество элементов в S_n:
$$ 
    \#S_n = n! 
$$

$ \cdot $ Действие подстановки $\sigma \in S_n $ на множестве удобно записывать:
$$ 
    \sigma = \begin{pmatrix}
        1 & 2 & 3 & ... & n \\
        i_1 & i_2 & i_3 & ... & i_n \\
    \end{pmatrix}, \text{ где } i_k = \sigma(i), k=1, 2, ..., n 
$$

$ \cdot $ Поскольку подстановки - взаимно однозначные соответствия, то они умножаются в соответствии с правилом композиции:
$$ 
    \sigma, \tau \in S_n \text{, то } \sigma \times \tau(i) = \sigma(\tau(i))
$$

$ \cdot $ *Нейтральным* элементом в группе подстановок является тождественная подстановка $ e(i) = i, i=1,2,3,...,n $.

$ \cdot $ *Обратным* является обратная подстановка (возвращающая элементы множества M на изначальные позиции) - $ \sigma^{-1}(i) $.

### 3) Циклы:

Подстановка называется циклом длины k, если на некотором подмножестве $ \{ a_{i_1}, a_{i_2}, ..., a_{i_k} \} $ множества M она действует следующим образом:
$$
    \begin{pmatrix}
        a_{i_1} & a_{i_2} & ... & a_{i_{k-1}} & a_{i_k} \\
        a_{i_2} & a_{i_3} & ... & a_{i_k} & a_{i_1} \\
    \end{pmatrix}, \text{ а остальные элементы M не меняются.}
$$

$ \cdot $ Обозначается цикл: $ (a_{i_1}, a_{i_2}, ..., a_{i_k}) $

$ \cdot $ Два цикла независимы, если они действуют на непересекающихся подмножествах M.

$ \cdot $ Пусть $ \sigma $ - цикл длины k. Тогда он порождает в циклическую подгруппу $ S_n $: $ <\sigma> = \{ e, \sigma, \sigma^2, ..., \sigma^{k-1} \} $ порядка k.

$ \cdot $ $ (\sigma^S)^{-1} = \sigma^{k-S} $, $ s = 1, 2, ..., k-1 $.

$ \cdot $ Теорема. Каждая подстановка $ \sigma \neq e $ в $ S_n $ является произведением независимых циклов. то разложение в произведение определено однозначно с точностью до порядка следования циклов.

### 4) Группа действий на множество:

Пусть $ G $ - группа с нейтральным элементом e, а $ X $ - множество. Будем говорить, что G действует на X, если задана операция $ G \times X -> G $, обладающая следующими свойствами (для $ \forall x \in X $ и $ g, h \in G $):
- e(x) = x
- $ g \times h(x) = g(h(x)) $

$ \cdot $ *Орбитой элемента* $ x \in X $ под действием $ G $ называется множество $ G_x = \{ gx: g \in G \} $. Число $ \#(G_x) $ называется длиной орбиты.

$ \cdot $ *Неподвижными точками* элемента $ g \in G $ называется $ x \in X $, для которых $ gx = x $. Множество неподвижных точек элемента $ g $ обозначают $ X^g $.

### 5) Лемма Бернсайда:

Количество орбит действия группы $ G $ на множестве $ X $ равно:
$$
    N = \frac{1}{\#G}\sum_{g \in G}{\#X^g}
$$

### 2. Графическое отображение раскраски по заданному номеру:

Реализуем функцию, которая позволяет визуализировать любую возможную раскраску бус:

```python
# Поворот вектора на угол:
def rotate_by_angle(x_y, angle):
    angle = angle * pi()/180
    rotation_mat = matrix([[cos(angle), -sin(angle)],[sin(angle), cos(angle)]])
    return (rotation_mat * x_y.transpose()).transpose()

# Отрисовка бус по номеру:
def print_beads(num_coloring, distance = 20, circle_size = 400, num_circles = 38):
    num_colors = 4
    colors = ["red", "green", "blue", "purple"]

    # Получение цветов:
    circles_color = []
    for _ in range(38):
        circles_color.append(colors[num_coloring % num_colors])
        num_coloring //= num_colors
    # Процесс построения :
    x_y = matrix([distance, 0])
    base_angle = 360 / num_circles
    beads = point(x_y.list(), color = circles_color[0], size = circle_size)
    for i in range(1, num_circles):
        curr_center_x_y = (rotate_by_angle(x_y, base_angle * i)).list()
        beads += point(curr_center_x_y, color = circles_color[i], size = circle_size)

    show(beads, axes=False)
```

### 3. Решение задачи в общем виде:

1) Сначала сформулируем саму задачу в общем виде:

Сколькими способами можно составить бусы из n бусинок k различных цветов (с точностью до поворотов и осевых симметрий)?

**I) Решим для четных n:**

2) Рассмотрим как измениться группа $ G^{*} $:

Во-первых, она станет подгруппой $ S_n $.

Во-вторых, все её перестановки будут теперь:

$$  e, \sigma, \sigma^{2}, \sigma^{3}, ..., \sigma^{n-1} $$
$$  \tau, \tau \times \sigma, \tau \times \sigma^{2}, \tau \times \sigma^{3}, ..., \tau \times \sigma^{n-1} $$

Причем $ \sigma $ (Поворот бус на 1 бусину против часовой стрелки) и $ \tau $ (осевая симметрия относительно оси, разделяющей $ \{ 1, 2, 3, ..., \frac{n}{2} \} $ и $ \{ \frac{n}{2} + 1, \frac{n}{2} + 2, ..., n  \} $) мы знаем сразу:

$ \sigma = (1, 2, 3, 4, ..., n) $ - перестановка имеет 1 независимый цикл длины n.

$ \tau = (1, n)(2, n-1)...(\frac{n}{2}, \frac{n}{2} + 1) $ - перестановка имеет $ \frac{n}{2} $ независимых циклов длины 2.

3) Для применения леммы Бернсайда (нахождения количества способов составить бусы с точностью до поворотов и осевых симметрий) нам необходимо лишь знать число циклов для всех перестановок $ G^{*} $. Число циклов для $ \sigma^{2}, \sigma^{3}, ..., \sigma^{n-1} $ мы можем получить из следующего соображения:

Если $ \sigma $ является циклом длины $ m $, а НОД(m, v) = l. Тогда перестановка $ \sigma^v $ является произведением $ l $ циклов длины $ \frac{m}{l} $.

Тогда количество циклов для $ \sigma^{v} = НОД(n, v) $ для $ v = 2, 3, ..., n-1 $.

4) Осталось разобраться с количеством циклов для $ \tau \times \sigma, \tau \times \sigma^{2}, ..., \tau \times \sigma^{n-1} $. Рассмотрим операции, которые мы производим над изначальным размещением бусин, мы сначала поворачиваем на определенное число бусин, а затем отражаем, причем возможны две ситуации: все бусины перешли в другие или 2 остались на месте, причем вторая ситуация происходит через раз, т.е. при применении $ \tau \times \sigma $ 2 бусины останутся на своих местах, при применении $ \tau \times \sigma^{2} $ все будут на разных позициях, $ \tau \times \sigma^3 $ снова 2 бусины будут на своих позициях. Это происходит потому что поворот на какое-то количество бусин происходит симметрично относительно оси симметрии, поэтому совпадающих бусин будет всегда четное число (при четном n). А всего их будет 2 потому что каждой бусине в одной из половин бус необходимо разное число поворотов, чтобы при отражении попасть обратно на свою позицию. Таким образом получаем:

Количество циклов для $ \tau \times \sigma^{v} = \frac{n}{2} + mod(v, 2) $ для $ v = 1, 2, 3, ..., n-1 $.

5) Осталось применить лемму Бернсайда и получим ответ:

$$
    N = \frac{1}{\#G}\sum_{g \in G}{\#X^g} = \frac{1}{2n} \left(k^{n} + k + \sum_{v = 2}^{n-1}{k^{НОД(n, v)}} + k^{\frac{n}{2}} + \sum_{v = 1}^{n-1}{k^{\frac{n}{2} + mod(v, 2)}} \right) 
$$

**I) Решим для нечетных n:**

2) Рассмотрим как измениться группа $ G^{*} $:

Во-первых, она станет подгруппой $ S_n $.

Во-вторых, все её перестановки будут теперь:

$$  e, \sigma, \sigma^{2}, \sigma^{3}, ..., \sigma^{n-1} $$
$$  \tau, \tau \times \sigma, \tau \times \sigma^{2}, \tau \times \sigma^{3}, ..., \tau \times \sigma^{n-1} $$

Причем $ \sigma $ (Поворот бус на 1 бусину против часовой стрелки) и $ \tau $ (осевая симметрия относительно оси, разделяющей $ \{ 1, 2, 3, ..., [\frac{n}{2}] + 1 \} $ и $ \{ [\frac{n}{2}] + 1, [\frac{n}{2}] + 2, ..., n  \} $) мы знаем сразу:

$ \sigma = (1, 2, 3, 4, ..., n) $ - перестановка имеет 1 независимый цикл длины n.

$ \tau = (1, n)(2, n-1)...([\frac{n}{2}], [\frac{n}{2}] + 2)([\frac{n}{2}] + 1) $ - перестановка имеет $ [\frac{n}{2}] + 1 $ независимых циклов длины 2.

3) Для применения леммы Бернсайда (нахождения количества способов составить бусы с точностью до поворотов и осевых симметрий) нам необходимо лишь знать число циклов для всех перестановок $ G^{*} $. Число циклов для $ \sigma^{2}, \sigma^{3}, ..., \sigma^{n-1} $ мы можем получить из следующего соображения:

Если $ \sigma $ является циклом длины $ m $, а НОД(m, v) = l. Тогда перестановка $ \sigma^v $ является произведением $ l $ циклов длины $ \frac{m}{l} $.

Тогда количество циклов для $ \sigma^{v} = НОД(n, v) $ для $ v = 2, 3, ..., n-1 $. (Не зависит от четности n).

4) Осталось разобраться с количеством циклов для $ \tau \times \sigma, \tau \times \sigma^{2}, ..., \tau \times \sigma^{n-1} $. Рассмотрим операции, которые мы производим над изначальным размещением бусин, мы сначала поворачиваем на определенное число бусин, а затем отражаем, причем в любом случае одна бусина останется на своем месте, а остальные перейдут друг в друга, так как ось симметрии проходит через бусину с номером $ [\frac{n}{2}] + 1 $ и получается так, что для каждой бусины в бусах необходимо разное количество поворотов, чтобы при отражении она попала обратно на свою позицию. Таким образом получаем:

Количество циклов для $ \tau \times \sigma^{v} = [\frac{n}{2}] + 1 $ для $ v = 1, 2, 3, ..., n-1 $.

5) Осталось применить лемму Бернсайда и получим ответ:

$$
    N = \frac{1}{\#G} \sum_{g \in G}{\#X^g} = \frac{1}{2n} \left(k^{n} + k + \sum_{v = 2}^{n-1}{k^{НОД(n, v)}} + \sum_{v = 1}^{n}{k^{[\frac{n}{2}] + 1}} \right) 
$$

В общем получаем:

n - четное:

$$
    N = \frac{1}{2n} \left(k^{n} + k + \sum_{v = 2}^{n-1}{k^{НОД(n, v)}} + k^{\frac{n}{2}} + \sum_{v = 1}^{n-1}{k^{\frac{n}{2} + mod(v, 2)}} \right) 
$$

n - нечетное:

$$
    N = \frac{1}{2n} \left(k^{n} + k + \sum_{v = 2}^{n-1}{k^{НОД(n, v)}} + \sum_{v = 1}^{n}{k^{[\frac{n}{2}] + 1}} \right) 
$$

Проверим полученную формулу для нашей задачи (можем проверить только формулу для четных n):

```python
n = 38
k = 4

N = (1 / (2*n))*((k ** n) + k + sum([k ** gcd(n, v) for v in range(2, n)]) + (k ** (n/2)) + sum([k ** (n/2 + (v % 2)) for v in range(1, n)]))
print(N)
```

Полностью совпадает с найденным нами ранее числом - 994182417793455295870.
