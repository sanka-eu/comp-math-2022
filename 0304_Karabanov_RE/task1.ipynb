–¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º –≤–∏–¥–Ω–æ, —á—Ç–æ –¥–ª—è —á–∏—Å–ª–∞ 33 –º–µ—Ç–æ–¥ –º–Ω–æ–∂–∏—Ç–µ–ª–µ–π –¥–∞—ë—Ç —Ä–µ–∑—É–ª—å—Ç–∞—Ç –ø–æ –æ–ø–µ—Ä–∞—Ü–∏—è –±–æ–ª—å—à–∏–π, —á–µ–º –±–∏–Ω–∞—Ä–Ω—ã–π –º–µ—Ç–æ–¥, —Ç.–∫ –æ–Ω —Ä–∞–±–æ—Ç–∞–µ—Ç "–º–µ–¥–ª–µ–Ω–Ω–µ" –ø–æ –æ—Ç–Ω–æ—à–µ–Ω–∏—é –∫ –±–∏–Ω–∞—Ä–Ω–æ–º—É –º–µ—Ç–æ–¥—É –ø—Ä–∏ —Å—Ç–µ–ø–µ–Ω—è—Ö $ n =  2^k + 1 $ –≤ —Ç–æ –≤—Ä–µ–º—è, –∫–∞–∫ –¥–ª—è —á–∏—Å–ª–∞ 125 –º–µ—Ç–æ–¥ –º–Ω–æ–∂–∏—Ç–µ–ª–µ–π –¥–∞—ë—Ç –º–µ–Ω—å—à–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –æ–ø–µ—Ä–∞—Ü–∏–π, –ø–æ —Å—Ä–∞–≤–Ω–µ–Ω–∏—é —Å –±–∏–Ω–∞—Ä–Ω—ã–º –º–µ—Ç–æ–¥–æ–º. –ú–æ–∂–Ω–æ —Å–¥–µ–ª–∞—Ç—å –≤—ã–≤–æ–¥—ã –æ —Ç–æ–º, —á—Ç–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –æ–ø–µ—Ä–∞—Ü–∏–π, –Ω–µ–æ–±—Ö–æ–¥–∏–º—ã—Ö –¥–ª—è –º–µ—Ç–æ–¥–∞ –º–Ω–æ–∂–∏—Ç–µ–ª–µ–π —Ä–∞–≤–Ω–æ —Å—É–º–º–µ –¥–ª–∏–Ω –∞–¥–¥–∏—Ç–∏–≤–Ω—ã—Ö —Ü–µ–ø–æ—á–µ–∫ —Å–æ–º–Ω–æ–∂–∏—Ç–µ–ª–µ–π, –ø–æ—ç—Ç–æ–º—É –Ω–∏ –æ–¥–∏–Ω –∏–∑ –º–µ—Ç–æ–¥–æ–≤ –Ω–µ –º–æ–∂–µ—Ç —Ä–∞—Å—Å–º–∞—Ç—Ä–∏–≤–∞—Ç—å—Å—è –∫–∞–∫ –±–æ–ª–µ–µ —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω—ã–π.

### –†–µ–∞–ª–∏–∑–∞—Ü–∏—è –∞–ª–≥–æ—Ä–∏—Ç–º–∞ –ë—Ä–∞—É—ç—Ä–∞
–†–µ–∞–ª–∏–∑–æ–≤–∞–Ω–Ω—ã–π –∞–ª–≥–æ—Ä–∏—Ç–º –ë—Ä–∞—É—ç—Ä–∞ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω, —Å –≤–∞—Ä–∏–∞—Ü–∏–µ–π –∑–Ω–∞—á–µ–Ω–∏–π k –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω –Ω–∏–∂–µ:

```python
def alg(n, k):
      d = 2 ** k
      ans = []
      if (n < d):
        i = 0
        while i < n:
            ans.append(i + 1)
            i += 1
      else:
          q = n
          ans_r = []
          while q >= d:
              r = q % d
              ans_r.append(r)
              q = q // d
          ans_r.append(q)
          ans_r.reverse()
          z = 1
          while z < d:
              ans.append(z)
              z += 1          
          q = ans_r[0]
          i = 0
          while i < (len(ans_r) - 1):
              j = 0
              while j < k:
                  q *= 2
                  ans.append(q)
                  j += 1
              if ans_r[i+1] > 0:
                  q += ans_r[i+1]
                  ans.append(q)
              i += 1
      return ans, (len(ans))

n = int(input("n: "))
k = int(input("k: "))
ad = []
lng = 0
ad, lng = alg(n, k)
print("–ê–¥–¥–∏—Ç–∏–≤–Ω–∞—è —Ü–µ–ø–æ—á–∫–∞: ", ad)
print("–î–ª–∏–Ω–∞ –∞–¥–¥–∏—Ç–∏–≤–Ω–æ–π —Ü–µ–ø–æ—á–∫–∏: ", lng)
```

### –ó–∞–¥–∞–Ω–∏–µ 3
### –ê–ª–≥–æ—Ä–∏—Ç–º –¥—Ä–æ–±–ª–µ–Ω–∏—è –≤–µ–∫—Ç–æ—Ä–∞ –∏–Ω–¥–µ–∫—Å–æ–≤


–û–ø–∏—à–µ–º —Ñ—É–Ω–∫—Ü–∏–∏ $ \nu(n) $ –¥–ª—è —É—Ç–æ—á–Ω–µ–Ω–∏—è –æ—Ç—Ä–µ–∑–∫–æ–≤ —Ä–∞—Å—Å–º–∞—Ç—Ä–∏–≤–∞–µ–º—ã—Ö –¥–ª–∏–Ω —Ü–µ–ø–æ—á–µ–∫:

```python
def nu(n):
    res = 0
    while n != 0:
        if n % 2 == 1:
            res += 1
        n //= 2

    return res
```

–î–∞–ª–µ–µ –æ–ø–∏—à–µ–º —Å–∞–º –∞–ª–≥–æ—Ä–∏—Ç–º –¥—Ä–æ–±–ª–µ–Ω–∏—è –∏–Ω–¥–µ–∫—Å–æ–≤ –∏ –ø–æ –Ω–µ–º—É –ø–æ—Å—Ç—Ä–æ–∏–º –º–∏–Ω–∏–º–∞–ª—å–Ω—É—é –∑–≤—ë–∑–¥–Ω—É—é —Ü–µ–ø–æ—á–∫—É

```python
import time

def get_a(vec): 
    additive_chain = []
    additive_chain.append(1)
    for i in range(len(vec)):
        additive_chain.append(additive_chain[i] + additive_chain[vec[i] - 1])
    
    return additive_chain[-1]

def reduce(vec):
    _len = len(vec)
    for i in range(_len):
        if vec[_len - i - 1] > 1:
            vec[_len - i - 1] -= 1
            for j in range(_len - i, _len):
                vec[j] = j + 1
            break

def get_star_chain(star_chain_vec):
    additive_chain = [1]
    for i in range(len(star_chain_vec)):
        additive_chain.append(additive_chain[i] + additive_chain[star_chain_vec[i] - 1])
    
    return additive_chain

def not_reduce(vec):
    for i in vec:
        if i > 1:
            return False
    
    return True

def splitting_alg(n):
    lower_border = math.ceil(math.log(n, 2))
    upper_border = int(math.log(n, 2) + nu(n))
    for m in range(lower_border, upper_border):
        q = int(m/2)
        fix_path = [i for i in range(1, q + 1)]
        while True:
            change_path = [q + i for i in range(1, m - q + 1)]
            a = get_a(fix_path)
            a_min = a + m - q
            a_max = a * 2 ** (m - q)

            if n < a_min or n > a_max:
                if not_reduce(fix_path) and len(fix_path) > 1:
                    break
                reduce(fix_path)
                continue
                
            while True: 
                if get_a(fix_path + change_path) == n:
                    return fix_path + change_path
                if not_reduce(change_path):
                    break
                reduce(change_path)
            if not_reduce(fix_path):
                break
            reduce(fix_path)   

n = int(input())
start = time.time()
star_chain_vec = splitting_alg(n)
end = time.time()
print("–ó–≤—ë–∑–¥–Ω–∞—è —Ü–µ–ø–æ—á–∫–∞ : ", get_star_chain(star_chain_vec))
print("–î–ª–∏–Ω–∞ –∑—Ü–µ–ø–æ—á–∫–∏ : ", len(get_star_chain(star_chain_vec)))
print("–í—Ä–µ–º—è –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è : ", end - start)
```
