## Основные теоретические положения
**Группа** - множество G с бинарной операцией * называется группой, если:  
1) операция * ассоциативна  
2) в G существует нейтральный элемент $e$  
3) $\forall a \in G \; \exists$ обратный элемент $a' \in G$, такой что $a'*a = a*a' = e$.  
Пусть $G$ - группа с нейтральным элементом $e$, а $X$ - множество. Будем говорить, что $G$ действует на $X$, если задана операция: $G\times X \rightarrow X$, обладающая для любого $x\in X$ и $g, h \in G$ следующими свойствами:  
1) $ex = x $  
2) $gh(x) = g(h(x))$  
**Орбита** - орбитой элемента $x \in X$ под действием $G$ называется множество $Gx = \{ gx : g \in G \}$. Число $\#(G_x)$ называется длиной орбиты.  
Неподвижными точками элемента $g \in G$ называют точки $x \in G$, для которых $gx = x$.  
Множество неподвижных точек элемента $g$ обозначают $X^g$  
### Лемма Бернсайда

Количество орбит N действия группы $G$ на множестве $X$ задается следующей формулой: <center> $ N = \dfrac{1}{\# G}\displaystyle\sum\limits_{g \in G} \#X^g $ </center>  
### Задача о раскрасках  
Пусть $I$ - произвольное множество, а C - множество цветов. Расскраской $i$ называется функция из $I$ в $C$.  
 $C^I$ - множество всех раскрасок.
 
 
 $\#C^I = \#C^{\#I}$  
 Пусть $g \in G, f \in C^I$. Тогда для этих $g$ и $f$ следующие условия эквивалентны:  
 1) $gf = f$  
 2) $f(i) = f(g^ni) \; \forall \; i \in I, \forall \; n \in Z$.  
 3) Если $i, j \in I$ лежат в одном цикле в циклической записи перестановки $\alpha(g)$, то $f(i) = f(g)$.  
Количество раскрасок из множетсва $C^I$, которые сохраняют элемент $g \in G$ равно $\#C^{c(\phi(g))}$, $c(\phi(g))$ - число независимых циклов перестановки $\phi(g)$

### Графическое отображение раскраски

```python
from sage.plot.polygon import polygon, polygon2d
import random


class Hexagon:
    side = 6 # размер стороны треугольников
    height = side * pow(3, 0.5) / 2 # высота
    colors = ['#6600ff', '#fde910', '#ff2400', '#00ff00'] #цвета
    edge_color = 'black'  # цвет рёбер
    parts = 6       # деление сторон

    def __init__(self, num):
        self.triangles = [x for x in range(4 * self.parts - 1, 4 * self.parts - 1 - self.parts * 2, -2)] #заполнение треугольников на уровнях
        self.num = num  # номер раскраски
        self.figure = self.create_hex(self.num)

    def make_triangle(self, x, y, color, reverse): # создание треугольника в зависимости от порядка
        dote_1 = (x, y) # первая вершина
        dote_2 = (x + self.side, y) if not reverse else (x - self.side / 2, y + self.height) # вторая вершина
        dote_3 = (x + self.side / 2, y + self.height) if not reverse else (x + self.side / 2, y + self.height) # третья вершина

        return polygon([dote_1, dote_2, dote_3], color=color, edgecolor=self.edge_color) # треугольник

    def create_half(self, triangles, num, start_orientation=True):
        hex_half = polygon([0, 0], axes=False) # создание начальных характеристик в зависимости от половины
        start_x = 0 if not start_orientation else self.side / 2 # стартовая точка
        start_y = 0 if not start_orientation else -1 * self.height # стартовая точка
        y_direction = 1 if not start_orientation else -1 # направление относительно y

        for triangle in triangles: #заполняем послойно треугольники в зависимости от уровня
            orientation = start_orientation
            x = start_x
            y = start_y

            for i in range(triangle): # кл-во треугольников на текущем уровне
                color = self.colors[num % len(self.colors)]  # подбираем цвет
                hex_half = hex_half + self.make_triangle(x, y, color, orientation) #формируем половину шестиугольника

                if not orientation: # смещаем координату по x
                    x = x + self.side

                orientation = not orientation # меняем треугольник на перевёрнутый
                num = num * 95 // 96 # формируем другой цвет

            start_x = start_x + self.side / 2 # переход на следующий уровень
            start_y = start_y + y_direction * self.height

        return hex_half, num

    def create_hex(self, num):
        hex_shape = polygon([0, 0], axes=False) # создаём систему координат

        half, new_num = self.create_half(self.triangles, num) # отрисовываем нижнюю половину

        hex_shape = hex_shape + half  + self.create_half(self.triangles, new_num, False)[0] # отрисовываем верхнюю половину

        return hex_shape

    def show(self):
        self.figure.show() # вывод на экран
```

###  Решение задачи в общем виде

Задача в общем виде имеет следующий вид: Каждая сторона правильного шестиугольника разбита на $m$ равных частей. Через точки деления и вершины проведены отрезки, параллельные сторонам шестиугольника. В получившиеся треугольные клеточки вставляют цветные стекла $k$ различных цветов. Сколькими способами можно составить такую мозаику (с точностью до поворотов и осевых симметрий шестиугольника).

Принцип решения задачи не поменяется, т.к. меняется только количество треугольников в шестиугольнике и количество цветов.

Таким образом по лемме Бернсайда формула количества раскрасок будет иметь следующий вид:

$N = \frac{1}{12}((k^{m^2})^{6} + 2*k^{m^2} + 2*(k^{m^2})^{2} + 4*(k^{m^2})^{3} + 3*(k^{m^2})^{4}) $,  
где k - кол-во цветов,  
m - кол-во частей, на которые разделён шестиугольник
