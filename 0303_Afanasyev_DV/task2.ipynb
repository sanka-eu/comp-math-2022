## Основные теоретические положения
__Действие группы на множество__

_Def:_ $\; \sqsupset G$ - группа с нейтральным элементом $e$, а $X$ - какое-то множество.

$G$ действует на $X$, если задана операция: $G\times X \rightarrow X$, будем обозначать $gx$, где $x\in X$, $g \in G$, 
которая обладает следующими свойства: ( $\forall x \in X$, $\forall g,h \in G$)

<ol>
<li> $ex = x$ </li>
<li> $g(hx) = (gh)x$ </li>
</ol>  

__Орбита элемента__

_Def:_ Орбита элемента $x \in X$ под действием группы $G$ - множество $Gx = \left\{gx \; \big| \; g \in G\right\}$.

Количество элементов в данной орбите, называется длиной орбиты.

__Неподвижные точки__

_Def:_ Неподвижные точки элемента $g \in G$ - $X^g$ множество $x \in X$, для которых выполняется $gx = x$.


__Лемма Бёрнсайда__
 
 _Theorem:_$\;$Количество орбит действия группы $G$ на множество $X$ равно: 
 <center> $ N = \dfrac{1}{\# G}\displaystyle\sum\limits_{g \in G} \#X^g $ </center>
 
 __Постановка задачи о раскрасках__
 
 _Def:_ $\sqsupset I$ - произвольное множество, $С$ - множество цветов. Раскраской множества $I$ называется функция $I \rightarrow C$
 
 $C^I$ - множество всех раскрасок.

$\#C^I = \#C^{\#I}$

 _Def:_ Пусть группа  G  действует на множество  I , тогда формула $gf(i)=f(g^{−1}(i)),g \in G,i \in I,f \in C^I$
задаёт действие $G$ на множество  $C^I$.

 _Def:_ Для элемента $g \in G$ и функции $f \in C^I$ следующие условия эквивалентны:
<ol>
<li>$gf = f$</li>

<li>$f(i) = f(g^{n}i)$, $\forall i \in I $ и $n \in \mathbb{Z}$</li>

<li>Если $i,j \in I$ лежат в одном цикле в циклической записи перестановки $\varphi(g)$, то $f(i) = f(j)$</li>
</ol>

 _Def:_ Количество раскрасок из множетсва $C^I$, которые сохраняют элемент $g \in G$ равно $\#C^{c(\phi(g))}$, $c(\phi(g))$ - число независимых циклов перестановки $\phi(g)$

**Реализация функции, отображающая работу функции графически раскраску по заданному номеру** 

```python
import math
import random

def gen_col(var):
    if(var==1):
        return "red"
    if(var==2):
        return "green"
    if(var==3):
        return "blue"
    if(var==4):
        return "white"
    if(var==5):
        return "black"
    if(var==6):
        return "gold"
    else:
        return "empty"

def paint_tetr(col):
    x=[]
    y=[]
    z=[]
    for i in range(4):
        x.append(0)
        y.append(0)
        z.append(0)

    z[0]=z[1]=0
    z[2]=0
    z[3]=2+1

    x[0]=2
    x[1]=-x[0]
    x[2]=x[0]+1
    x[3]=x[2]+2

    y[0]=1
    y[1]=2
    y[2]=3
    y[3]=2


    Pol1=polygon3d([(x[0],y[0],z[0]), (x[1],y[1],z[1]), (x[2],y[2],z[2])], color = gen_col(col[0]))
    Pol2=polygon3d([(x[0],y[0],z[0]), (x[1],y[1],z[1]), (x[3],y[3],z[3])], color = gen_col(col[1]))
    Pol3=polygon3d([(x[3],y[3],z[3]), (x[1],y[1],z[1]), (x[2],y[2],z[2])], color = gen_col(col[2]))
    Pol4=polygon3d([(x[0],y[0],z[0]), (x[3],y[3],z[3]), (x[2],y[2],z[2])], color = gen_col(col[3]))

    
    return Pol1 + Pol2 + Pol3 + Pol4

col=[]
for i in range(4):
    col.append(random.randint(1,6))
graph = paint_tetr(col)
graph.show()
```
